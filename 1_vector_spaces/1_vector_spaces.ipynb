# Fields

## Definition

In the scope of this book:

Let $K$ be a subset of the complex numbers $C$. $K$ is a **field** if it satisfied the following conditions:
- (a) If $x, y$ are elements of $K$, then $x + y$ and $x y$ are also elements of $K$.
- (b) If $x \in K$, then $- x$ is also an element of $K$. If furthermore $x \neq 0$, then $x^{- 1}$ is an element of $K$.
- (c) The element $0$ and $1$ are elements of $K$.

Note: there is a [more abstract definition of a field](https://en.wikipedia.org/wiki/Field_(mathematics)#Classic_definition).

## Subfields

If $K, L$ are fields, and $K \subseteq L$, then $K$ is a **subfield** of $L$.

Hence in the scope of this book, all fields are subfields of the complex numbers $C$.

## Numbers/Scalars

An element of a field is call a **number** or a **scalar**.

# Vector Spaces

## Definition

A set $V$ is a vector space over the field $K$ if:

- Definition of operations: For $v ,w \in V$, and $c \in K$
 - Sum between two vectors ($v + w$) is defined.
 - Multiplication of a vector by a number (scalar multiplication) ($c v$) is defined.
- All of the specific vector spaces properties holds:
 - **VS 1.** (Additive associativity) Let $u, v, w \in V$, we have
 $$
 (u + v) + w = u + (v + w)
 $$
 - **VS 2.** (Additive identity) There is an element of $V$, denoted by $\mathit{0}$, such that for all element $u \in V$
 $$
 \mathit{0} + u = u + \mathit{0} = u
 $$
 Typesetting note: note the difference of $0$ (`$0$`), the additive identity in a field, and $\mathit{0}$ (`$\mathit{0}$`), the additive identity in a vector space.
 - **VS 3.** (Additive inverse) For $u \in V$, there exists an element in $V$
 $$
 u + (- u) = \mathit{0}
 $$
 - **VS 4.** (Commutativity) For all elements $u, v \in V$, we have
 $$
 u + v = v + u
 $$
 - **VS 5.** (Distributivity of scalar multiplication with respect to vector addition) Let $c$ be a number ($c \in K$), and $u, v \in V$ we have
 $$
 c (u + v) = c u + c v
 $$
 - **VS 6.** (Distributivity of scalar multiplication with respect to field addition) Let $a, b$ be numbers ($a, b \in C$), for $v \in V$, we have
 $$
 (a + b) v = a v + b v
 $$
 - **VS 7.** (Compatibility of scalar multiplication with field multiplication) Let $a, b$ be numbers ($a, b \in C$), for $v \in V$, we have
 $$
 (a b) v = a (b v)
 $$
 - **VS 8.** (Identity element of scalar multiplication) $1$ the additive identity element in $K$. For all elements $u \in V$, we have $1 \cdot u = u$.

# Linear Combinations

Let $V$ be an arbitrary vector space over a field $K$, and let $v_1, ..., v_n \in V$. Let $x_1, ..., x_n \in K$. Then

$$
x_1 v_1 + ... + x_n v_n
$$

is called a **linear combination** of $V$.

# Subspaces

Let $V$ be a vector space over a field $K$, and let $W \subseteq V$. $W$ is defined as a **subspace** of $V$ if $W$ satisfies all of the following conditions
- (i) Let $v, w \in W$, $(v + w) \in W$ also holds.
- (ii) Let $v \in W$, and $c \in K$, $c v \in W$ also holds.
- (iii) $\mathit{0} \in V$ holds.

# Generation / Linear Span

If every element of $V$ is a linear combination of $v_1, ..., v_n$, then we say that $v_1, ..., v_n$ **generates** (spans) $V$.

# Dot Products / Scalar Products

Let $A = (a_1, .., a_n), B = (b_1, ..., b_n) \in K^n$, the **dot product** (or scalar product) of $A$ and $B$ as

$$
A \cdot B = a_1 b_1 + ... + a_n, b_n
$$

$A \cdot A$ can be written as $A^2$.

$A^3$ has no meaning.

## Properties of dot products
- **SP 1.** Let $A, B \in K^n$, then we have
  $$
  A \cdot B = B \cdot A
  $$
 - Can be verified by multiplicative commutativity of $K$.
- **SP 2.** Let $A, B, C \in K^n$, then we have
  $$
  A \cdot (B + C) = A \cdot B + A \cdot C = (B + C) \cdot A
  $$
 - Can be verified by multiplicative commutativity and distributivity of $K$.
- **SP 3.** If $x \in K$, then
  $$
  (x A) \cdot B = x (A \cdot B) 
  $$
  and
  $$
  A \cdot (x B) = x (A \cdot B)
  $$
  - Can be verified by distributivity of $K$.

## Perpendicular / Orthogonal

Let $A, B \in K^n$. $A$ and $B$ are **perpendicular** (or, orthogonal) if $A \cdot B = 0$.

# Function Spaces

Let $S$ be a set and $K$ a field. By a function of $S$ into $K$ we shall mean an association which to each element of $S$ associates a unique element of $K$. Thus if $f$ is a function of $S$ into $K$, we express this by the symbols

$$
f: S \to K
$$

We also say that f is a $K$-valued function.

The **sum** operation of two $K$-valued functions $f, g$, $f + g$, is defined as a $K$-valued function such that

$$
(f + g)(x) = f(x) + g(x)
$$ for $x \in S$

The scalar multiplication of a $K$-valued function $f$ and a number $c \in K$ is defined as a $K-valued$ function such that

$$
(c f) = c f(x)
$$ for $x \in S$

As now because
1. Both the **sum** operation of two $K$-valued functions, and the scalar multiplication are defined; 
2. $f(x) \in K$ for $x \in S$, hence VS 1 through 8 applies for the set of $K$-valued functions

From above, let $V$ be the function of all $K$-valued functions. Then $V$ is a vector space over K.

## Examples

Let $V$ be the set of all functions of $R$ into $R$. Then $V$ is a vector space over R.

Let $W$ be the subset of $V$ that includes all continuous functions of $R$ into $R$, then $W$ is a vector space since
- Let $f, g$ be continuous functions. Then we know $f + g$ is also a continuous function.
- Let $c \in R$, and $f$ be a continuous function. $c g$ is also a continuous function.
- $\mathit{0}(x) = 0$ is a continuous function of $R$ into $R$.

Let $U$ be the subset of $V$ that includes all differentiable functions of $R$ into $R$, then $U$ is a vector space.

# Linear Independence

Let $V$ be a vector space over $K$, and let $v_1, ..., v_n \in V$. $v_1, ..., v_n$ is **linear dependent** if there exists $a_1, ..., a_n \in K$ not all equal to $0$ such that

$$
a_1 v_1 + ... a_n v_n = \mathit{0}
$$

If there do not exist such numbers in $K$, which means if $a_1, ..., a_n$ are all $0$ for above, then we say $v_1, ..., v_n$ are **linear independent**.

# Bases

If elements $v_1, ..., v_n$ of vector space $V$ **generate** $V$, and in addition are linear independent, then we call $\{v_1, ..., v_n\}$ is a **basis** of $V$.

We should also say elements $v_1, ..., v_n$ **constitutes** a form of **basis** of $V$.

## Coordinates

To define the **coordinates** of an element $v \in V$ with respect to a basis, we need to introduce the following theorem

### Theorem 2.1.

Let $V$ be a vector space. Let $v_1, ..., v_n \in V$ be linear independent. Let $x_1, ..., x_n$ and $y_1, ..., y_n$ be numbers. Suppose that we have

$$
x_1 v_1 + ... + x_n v_n = y_1 v_1 + ... + y_n v_n
$$

Then $x_i = y_i$ for $i$ in $1, ..., n$.

*Proof.* Subtracting the right-hand side from the left-hand side, we get

$$
\begin{array}{l}
(x_1 v_1 + ... + x_n v_n) - (y_1 v_1 + ... + y_n v_n) &= 0 \\
(x_1 - y_1) v_1 + ... + (x_n - y_n) v_n &= 0
\end{array}
$$

Since $v_1, ..., v_n$ are linear independent, we have $x_i - y_i = 0$ for $i$ in $1, .., n$.

Hence $x_i = y_i$ for $i$ in $1, ..., n$. Q.E.D.

From **Theorem 2.1.** we know a vector can only be written as a unique form of linear combination of a basis of the vector space.

Let $v$ in vector space $V$. Let $\{v_1, ..., v_n\}$ be a basis of $V$. If $v$ can be written as the following form of the linear combination of the basis (let $x_1, ..., x_n$ be numbers):

$$
v = x_1 v_1 + ... + x_n v_n
$$

Then we call $(x_1, ..., x_n)$ the **coordinates** of $v$ with respect to the basis  $\{v_1, ..., v_n\}$.

## A criterion to determine when a set of elements of a vector space is basis

### Definition of maximal subset of linear independent elements

Let $\{v_1, ..., v_n\}$ be a set of elements of vector space $V$. Let $r$ be a positive integer $\leq n$. We shall say that $\{v_1, ..., v_r\}$ is a **maximal** subset of linear independent elements if $v_1, .., v_r$ if
- $v_1, ..., v_r$ are linear independent, and,
- Given any $v_i$ for with $i > r$, the elements $v_1, ..., v_r, v_i$ are linear dependent.

### Theorem 2.2.

Let $\{v_l, ..., v_n\}$ be a set of generators of a vector space V. Let $\{v_1, ..., v_r\}$ be a maximal subset of linearly independent elements. Then $\{v_l, ..., v_r\}$ is a basis of V.

*Proof.* First we need to prove $\{v_l, ..., v_r\}$ generates $V$.

To prove that, given $v_1, ..., v_n$ generate $V$, we shall prove that each $v_i$ for $i > r$ is a linear combination of $v_1, ..., v_r$.

Given $\{v_1, ..., v_r\}$ is the maximal set of linear independent elements, $v_1, ..., v_r, v_i$ linear dependent. Hence there exist numbers $x_1, ..., x_n, y$ that are not all $0$ such that

$$
x_1 v_1 + ... + x_n v_n + y v_i = 0
$$

And furthermore, $y \neq 0$, because $v_1, ..., v_n$ are linear independent, which contradicts with the condition.

Then we can rewrite the equation above as

$$
v_i = \frac{x_1}{- y} v_1 + ... + \frac{x_n}{- y} v_n
$$

for $i$ in $i > r$.

Hence $v_i$ for $i > r$ is a linear combination of $v_1, ..., v_n$. 

And since $v_1, ..., v_r, ..., v_n$ generate $V$, we have proven $\{v_l, ..., v_r\}$ generates $V$.

And since $v_1, ..., v_r$ are linearly independent, we have $\{v_1, ..., v_r\}$ is a basis of $V$.

## Theorem 3.1.

Let $V$ be a vector space over the field $K$. Let $\{v_1, ..., v_m\}$ be a basis of $V$ over $K$. Let $w_1, ..., w_n$ be elements of $V$, and assume that $n > m$. Then $w_1, ..., w_n$ are linearly dependent.

*Proof.* Assume $w_1, ..., w_n$ are linearly independent.

Since $\{v_1, ..., v_m\}$ is a basis of $V$, with some renumbering of $v_1, ..., v_m$, there exists elements $a_1, ..., a_m \in K$ such that

$$
w_1 = a_1 v_1 + ... + a_m v_m
$$

And by the assumption that $w_1, ..., w_n$ are linearly independent, we know $w_1 \neq 0$. Hence some $a_i \neq 0$. Without loss of generality, we say $a_1 \neq 0$. Then we have

$$
\begin{array}{l}
a_1 v_1 &= w_1 - a_2 v_2 - ... - a_m v_m \\
v_1 &= a_1^{-1} w_1 - a_1^{-1} a_2 v_2 - ... - a_1^{-1} a_m v_m
\end{array}
$$

As $v_1$ can be written as a linear combination of $w_1, v_2, ..., v_m$, and since $v_1, ..., v_m$ generate $V$, we have $w_1, v_2, ..., v_m$ generate $V$ as well.

Hence for $w_2 \neq 0$ there exists elements $a_1, ..., a_m \in K$ such that

$$
w_2 = a_1 w_1 + a_2 v2 + ... + a_m v_m
$$

Hence some $a_i \neq 0$. Without loss of generality, we say $a_2 \neq 0$. Then we have

$$
\begin{array}{l}
a_2 v_2 &= w_2 - a_1 w_1 - ... - a_m v_m \\
v_2 &= a_2^{-1} w_2 - a_2^{-1} a_1 w_1 - a_2^{-1} a_3 v_3 - ... - a_2^{-1} a_m v_m
\end{array}
$$

If we repeat the above steps, for $1 \leq r < m$, we will have

$$
w_{r + 1} = a_1 w_1 + ... + a_r w_r + a_{r + 1} v_{r + 1} +... + a_m v_m
$$

$$
\begin{array}{l}
a_{r + 1} v_{r + 1} &= w_r - a_1 w_1 - ... - a_r w_r - a_{r + 2} v_{r + 2} - ... - a_m v_m \\
v_{r + 1} &= a_{r + 1}^{-1} w_{r + 1} - a_{r + 1}^{-1} a_1 w_1 - ... a_{r + 1}^{-1} a_{r} v_{r} - a_{r + 1}^{- 1} a_{r + 2} v_{r + 2} - ... - a_{r + 1}^{-1} a_m v_m
\end{array}
$$

And eventually, we will have $w_1, ..., w_m$ generate $V$.

And also because $w_1, ..., w_n$ are linear independent by assumption, $w_1, ..., w_m$ are linear independent. Hence there exists elements $d_1, ... d_m$ in $K$ which are not zero such that

$$
w_{m + 1} = d_1 w_1 + ... d_m w_m
$$

which contracdicts the assumption that $w_1, ..., w_n$ are linear independent. Hence $w_1, ..., v_n$ are linear dependent. Q.E.D.

## Theorem 3.2.

Let V be a vector space and suppose that one basis has n elements, and another basis has m elements. Then m = n.

*Proof.* We apply **Theorem 3.1** to the two bases. **Theorem 3.1** implies that both alternatives $n > m$ and $m > n$ are impossible, and hence $m = n$. Q.E.D.

Now we have shown all bases in a vector space has the same number (cardinality) of elements.

## Definition of dimensions

Let $V$ be a vector space having a basis consisting of $n$ elements. We shall say that n is the **dimension** of $V$. If $V$ consists of $\mathit{0}$ alone, then $V$ does not have a basis, and we shall say that $V$ has dimension $0$.

A vector space which has a basis consisting of a finite number of elements, or the zero vector space, is called **finite dimensional**. Other vector spaces are called **infinite dimensional**. It is possible to give a definition for an infinite basis. The reader may look it up in a more advanced text. In this book, whenever we speak of the dimension of a vector space in the sequel, it is assumed that this vector space is finite dimensional.

## Theorem 3.3.

Let $V$ be a vector space, and $\{v_1, ..., v_n\}$ a maximal set of linearly independent elements of $V$. Then $\{v_1, ... ,v_n\}$ is a basis of $V$.

*Proof.*

In order to show $\{v_1, ... ,v_n\}$ is a basis of $V$, as we already know $v_1, ..., v_n$ are linear independent, we need to show $v_1, ... ,v_n$ generate $V$.

Let $w$ be any element of $V$. Since $\{v_1, ..., v_n\}$ is a maximal set, then $w, v_1, ... v_n$ must be linearly dependent. Hence there exists numbers $x_0, x_1, ..., x_n$ which are not all $0$ such that

$$
x_0 w + x_1 v_1 + ... + x_n v_n = \mathit{0}
$$

And $x_0$ cannot be $0$, otherwise, since $v_1, ..., v_n$ are linearly independent, we will obtiain a linear independent relation which leads to a contradiction. Then we have

$$
w = - x_0^{-1} x_1 v_1 - ... - x_0^{-1} x_n v_n
$$

Hence $v_1, ..., v_n$ generates $V$. Hence $\{v_1, ..., v_n\}$ is a basis. Q.E.D.

## Theorem 3.4.

Let $V$ be a vector space of dimension $n$, and let $v_1, ..., v_n$ be linearly independent elements of $V$. Then $v_1, ..., v_n$ constitute a basis of $V$.

*Proof.*

We need to show $v_1, ..., v_n$ genearate $V$.

According to **theorem 3.1**, there does not exist an $m > n$ such that $w_1, ..., w_m$ be linearly independent elements of $V$. Hence $v_1, ..., v_n$ is a *maximal subset of linear independent elements* of $V$.

Hence, according to **theorem 3.3** $v_1, ..., v_n$ constitute a basis of $V$. Q.E.D.

## Corollary 3.5.

Let $V$ be a vector space and let $W$ be a subspace. If $\dim W = \dim V$ then $V = W$.

*Proof.*

Since $\dim W = \dim V$, a basis of $W$ has the same number of element as of $V$. And since $W$ is a subspace of $V$, each of the element in a basis of $W$ is also an element of $V$. Then, according to *theorem 3.4*, a basis of $W$ is also a basis of $V$, which generates $V$ which lead to every element in $V$ is also in $W$.

And since $W$ is a subspace of $V$, each element in $W$ is also in $V$.

Hence $V = W$. Q.E.D.

## Corollary 3.6.

Let $V$ be a vector space of dimension $n$. Let $r$ be a positive integer with $r < n$, and let $v_1, ..., v_r$ be linearly independent elements of $V$. Then one can find elements $v_{r + 1}, ... ,v_n$ such that

$$
\{v_1, ..., v_n\}
$$

is a basis of $V$.

*Proof.*

As $\dim V = n$, the length of a maximal subset of linearly independent elements is $n$. And since $r < n$, $v_1, ..., v_r$ must not be a maximal subset of linearly independent elements. Hence there must exists $v_r \in V$ such that

$$
v_1, ..., v_r, v_{r + 1}
$$

...are linearly independent.

By induction, we can construct

$$
v_1, ..., v_r, ..., v_n
$$

...which are linearly independent and have the length of $n$. Q.E.D.

## Theorem 3.7.

Let $V$ be a vector space having a basis consisting of $n$ elements. Let $W$ be a subspace which does not consist of $\mathit{0}$ alone. Then $W$ has a basis, and $\dim W \leq n$.

*Proof.*

As $w$ is a subspace of $V$ does not contain $\mathit{0}$ alone, we can let $v$ be an element in $W$, and $v_1 \neq \mathit{0}$.

We then can treat $\{v_1\}$ as a linearly independent set of $W$.

### If $\{v_1\}$ is a maximal subset of linearly independent elements of $W$

...then by **theorem 3.3** $\{v_1\}$ is a basis of $W$, and $\dim W = 1$. Since $V$ has a basis, by definition of a basis we know $\dim V > 0$. Hence $\dim W \leq \dim V$.


### If $\{v_1\}$ is not a maximal subset of linearly independent elements of $W$

...then by induction, we can find a list of vectors $v_2, ..., v_r \in W$

$$
v_1, v_2, ..., v_r
$$

... as the maximal linearly independent subset of $W$. Then by **theorem 3.3**, $\{v_1, v_2, ..., v_r\}$ is a basis of $W$. Then $\dim W = r$.

And since $W$ is a subspace of $W$, $v_1, v_2, ..., v_r$ are linearly independent elements in $V$. If $r > n$, then by **theorem 3.1**, $v_1, v_2, ..., v_r$ is linearly dependent, which leads to a contradiction. Hence $r \leq n$ must be true.

Q.E.D.

# Sums and Direct Sums

## Definition of the Sum between two subspaces

Let $V$ be a vector space over the field $K$. Let $U, W$ be subspaces of $V$. We define the sum of $U$ and $W$ to be the subset of $V$ consisting of all sums $u + w$ with $u \in U$ and $w \in W$. We denote this sum by $U + W$.

## A sum is a subspace

### To prove property (i)

From the defnition of direct sums, let $u_1, u_2 \in U$, and $w_1, w_2 \in W$, then

$$
\begin{equation}
\begin{aligned}
(u_1 + w_1) + (u_2 + w_2) &= \underbrace{(u_1 + u_2)}_{\in U} + \underbrace{(w_1 + w_2)}_{\in W} &\text{by VS 1 of $V$} \\
&\in U + W
\end{aligned}
\end{equation}
$$equation

### To prove property (ii)

From the defnition of direct sums, let $u \in U$, $w \in W$, and $c \in K$ then

$$
\begin{equation}
\begin{aligned}
c (u + w) &= \underbrace{c u}_{\in U} + \underbrace{c w}_{\in W} &\text{by VS 5 of $V$} \\
&\in U + W
\end{aligned}
\end{equation}
$$

### To prove property (iii)

Since $U, W$ are subspaces of vector space $V$, we have

$$
\begin{equation}
\begin{aligned}
\mathit{0} &= \underbrace{\mathit{0}}_{\in U} + \underbrace{\mathit{0}}_{\in W} &\text{by VS 2 of $V$} \\
&\in U + W
\end{aligned}
\end{equation}
$$

Hence $U + W$ is a subspace of $V$. Q.E.D.

## Definition of a Direct Sums

$V$ is a **direct sum** of vector spaces $U$ and $W$ if for every element $v \in V$, there exists *unique* elements $u \in U$ and $w \in W$ such that $v = u + w$.

### Theorem 4.1.

Let $V$ be a vector space over the field $K$, and let $U, W$ be subspaces of $V$. If $U + W = V$, and if $U \cap W = \{\mathit{0}\}$, then $V$ is the direct sum of $U$ and $W$. We denote $V = U \oplus W$.

*Proof.*

For $v \in V$, since $U + W = V$, there exists $u \in U$ and $w \in W$ such that $v = u + w$.

To prove $V$ is the direct sum of $U$ and $W$, we need to show that $u$ and $w$ are unique.

Let there be $u^{\prime} \in U$ and $w^{\prime} \in W$ such that $v = u^{\prime} + w^{\prime}$.

Then we have

$$
\begin{array}{c}
u + w &= u^{\prime} + w^{\prime} \\
\underbrace{u - u^{\prime}}_{\in U} &= \underbrace{w^{\prime} - w}_{\in W}
\end{array}
$$

Then we know that $u - u^{\prime} = w^{\prime} - w$ and they belong to $U \cap W$.

Since $U \cap W = \{\mathit{0}\}$, we have

$$
\begin{cases}
u - u^{\prime} &= \mathit{0} \\
w^{\prime} - w &= \mathit{0}
\end{cases}
$$

Hence

$$
\begin{cases}
u &= u^{\prime} \\
w &= w^{\prime}
\end{cases}
$$

Hence  every element $v \in V$, there exists *unique* elements $u \in U$ and $w \in W$ such that $v = u + w$. Q.E.D.

### Theorem 4.2.

Let $V$ be a finite dimensional vector space over the field $K$. Let $W$ be a subspace. Then there exists a subspace $U$ such that $V = W \oplus U$.

*Proof.*

Since $V$ is a finite dimensional vector space, it has a basis with finite number of elements. Let $\dim V = n$ and $\{v_1, ..., v_n\}$ be a basis of $V$.

Since $W$ is a subspace of $V$, by **theorem 3.7**, we know $W$ has a subspace, and $\dim W < \dim V$. Let $\dim W = r$ and $\{w_1, ..., w_r\}$ be a subspace of $W$. We know $r < n$.

Since $\{w_1, ..., w_r\}$ is a basis of $W$, we have $w_1, ..., w_n$ are linearly independent elements in $W$, hence $w_1, ..., w_r$ are linearly independent elements in $V$. By **corollary 3.6**, there exist linearly independent elements $u_{r + 1}, ..., u_{n} \in V$ such that $\{w_1, ..., w_r, u_{r + 1}, ..., u_n\}$ is a basis of $V$.

Let $U$ be the vector space whose basis is $\{u_{r + 1}, ..., u_n\}$. 

Since $\{w_1, ..., w_r, u_{r + 1}, ..., u_n\}$ is a basis of $V$, then for each $v \in V$, there exist unique $a_1, ..., a_n$ such that

$$
\begin{equation}
\begin{aligned}
v &= \underbrace{a_1 w_1 + ... + a_r w_r}_{\in W} + \underbrace{a_{r + 1} u_{r + 1} + ... + a_n u_n}_{\in U}
\end{aligned}
\end{equation}
$$

Hence each $v \in V$, there are a unique pair of elements in $w \in W$ and $u \in U$ such that $v = u + w$.

Hence $V = W \oplus U$. Q.E.D.

### Theorem 4.3.

If $V$ is a finite dimensional vector space over $K$, and is the direct sum of subspaces $U, W$ then

$$
\dim V = \dim U + \dim W
$$

*Proof.*

Since $V$ is a finite dimensional vector space, and $U, W$ are subspaces of $V$, by **theorem 3.7** we know both $U, W$ have a basis.

Let $\{u_1, ..., u_r\}$ be a basis of $U$, and $\{w_1, ..., w_s\}$ be a basis of $W$.

Since $V = U \oplus W$, for each $v \in V$, there is a unique pair of $u \in U$ and $w \in W$ such that $v = u + w$.

Hence there are unique $x_1, ..., x_r$ and $y_1, ..., y_s$ such that

$$
v = x_1 u_1 + ... + x_r u_r + y_1 w_1 + ... + y_s w_s
$$

Hence $\{u_1, ..., u_r, w_1, ..., y_s\}$ is a basis of $V$. Hence $\dim V = r + s = \dim U + \dim W$. Q.E.D.

# Direct Product

## Definition

Suppose now that $U, W$ are arbitrary vector spaces over the field $K$ (i.e. not necessarily subspaces of some vector space). We let $U \times W$ be **the set of all pairs $(u, w)$** whose first component is an element $u$ of $U$ and whose second component is an element $w$ of $W$. We define the addition of such pairs componentwise, namely, if $(u_1, w_1) \in U \times W$ and $(u_2, w_2) \in U \times W$ we define

$$
(u_1, w_1) + (u_2, w_2) = (u_1 + u_2, w_2 + w_2)
$$

If $c \in K$ we define the product $c(u_1 , w_1)$ by

$$
c(u_1, w_1) = (c u_1, c w_1)
$$

Now, as the addition and multiplication have been defined, and the fact that $U, W$ are vectors spaces hence V1...V8 holds for them, we can validate that $U \times W$ is also a vector space.

$U \times W$ is called the **direct product** of $U$ and $W$.

## Dimension of direct products

$$
\dim (U \times W) = \dim U + \dim W
$$

*Proof.*

Let $n = \dim U$ and $m = \dim W$.

Let $u_1, ..., u_n$ be a basis of $U$ and $w_1, ..., w_m$ be a basis of $W$.

Let $\mathit{0_U}$ be the zero vector of $U$, and $\mathit{0_W}$ be the zero vector of $W$.

Let's construct a set $\{(u_1, \mathit{0_W}), ..., (u_n, \mathit{0_W}), (\mathit{0_U}, w_1), ..., (\mathit{0_U}, w_m)\}$. This set has $n + m$ number of elements.

Since $u_1, ..., u_n$ span $U$ and $w_1, ..., w_m$ span $W$, it is obvious that $(u_1, \mathit{0_W}), ..., (u_n, \mathit{0_W}), (\mathit{0_U}, w_1), ..., (\mathit{0_U}, w_m)$ span $U \times W$.

Now we need to show $(u_1, \mathit{0_W}), ..., (u_n, \mathit{0_W}), (\mathit{0_U}, w_1), ..., (\mathit{0_U}, w_m)$ are linearly independent. Since $u_1, ..., u_n$ are linearly independent, there does not exist $a_1, ..., a_n \in K$ which are not all $ = 0$ such that

$$
a_1 u_1, ..., a_n u_n = 0
$$

Likewise, there does not exist $b_1, ..., b_m \in K$ which are not all $ = 0$ such that

$$
b_1 w_1, ..., b_m w_m = 0
$$

Hence there does not exist $c_1, ..., c_{m + n}$ which are not all $ = 0$ such that

$$
c_1 (u_1, \mathit{0_W}) + ... + c_n (u_1, \mathit{0_W}) + c_{n + 1} (\mathit{0_U}, w_1) + ... + c_{n + m} (\mathit{0_U}, w_m) = 0
$$

Hence $(u_1, \mathit{0_W}), ..., (u_n, \mathit{0_W}), (\mathit{0_U}, w_1), ..., (\mathit{0_U}, w_m)$ are linearly independent.

Hence $\{(u_1, \mathit{0_W}), ..., (u_n, \mathit{0_W}), (\mathit{0_U}, w_1), ..., (\mathit{0_U}, w_m)\}$ is a basis of $U \times W$, and $\dim U \times W = n + m = \dim U + \dim W$. Q.E.D.
