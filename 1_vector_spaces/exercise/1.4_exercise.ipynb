## 1.  Let $V = R^2$, and let $W$ be the subspace generated by $(2, 1)$. Let $U$ be the subspace generated by $(0, 1)$. Show that $V$ is the direct sum of $W$ and $U$. If $U^{\prime}$ is the subspace generated by $(1, 1)$, show that $V$ is also the direct sum of $W$and $U^{\prime}$.

### Proof: Show that $V$ is the direct sum of $W$ and $U$.

The zero vector in $R^2$ is $\mathit{0} = (0, 0)$.

Since the first element of $(2, 1)$ is not $0$, and the second element of $(0, 1)$ is not $0$, $R^2 = W + U$.

Let $x, y \in R$ such that

$$
x (0, 1) + y (2, 1) = (0, 0)
$$

...which is identical to solve $x, y$ in this system of equations

$$
\begin{cases}
2 y &= 0 \\
x + y &= 0
\end{cases}
$$

...and the only solution to the system of equations is

$$
\begin{cases}
x &= 0 \\
y &= 0
\end{cases}
$$

Hence $W \cap U = \{\mathit{0}\}$.

By **theorem 4.1**, $V = R^2 = W \oplus U$ .

### Proof: Show that $V$ is the direct sum of $W$ and $U^{\prime}$.

Since the first element of $(2, 1)$ is not $0$, and the second element of $(1, 1)$ is not $0$, $R^2 = W + U^{\prime}$.


Let $x, y \in R$ such that

$$
x (2, 1) + y (1, 1) = (0, 0)
$$

...which is identical to solve $x, y$ in this system of equations

$$
\begin{cases}
2 x + y &= 0 \\
x + y &= 0
\end{cases}
$$

...and the only solution to the system of equations is

$$
\begin{cases}
x &= 0 \\
y &= 0
\end{cases}
$$

Hence $W \cap U^{\prime} = \{\mathit{0}\}$.

By **theorem 4.1**, $V = R^2 = W \oplus U^{\prime}$ .

Q.E.D.

## 3. Let $A, B$ be two vectors in $R^2$, and assume neither of them is $\mathit{0}$. If there is no number $c$ such that $c A = B$, show that $A, B$ form a basis of $R^2$, and that $R^2$ is a direct sum of the subspaces generated by $A$ and $B$ respectively.

### Proof: show that $A, B$ form a basis of $R^2$

First, we need to show that $A, B$ are linearly independent.

Assume $A$ $B$ are linearly dependent. Without loss of generality, there exist $a, b$ where $b \neq 0$ such that

$$
\begin{array}{c}
a A + b B = \mathit{0}
\frac{a}{b} A + B = \mathit{0}
- \frac{a}{b} A = B
\end{array}
$$

Hence there is a number $c = - \frac{a}{b}$ such that $c A = B$, which leads to a contradiction. Hence $A$ and $B$ are linearly dependent.

Then we need to show $A$ and $B$ generate $R^2$.

By **theorem 3.4**, since $\dim R^2 = 2$, and $A$ and $B$ are linearly independent, then $A$ and $B$ form a basis of $R^2$.

### Proof: show that $R^2$ is a direct sum of the subspaces generated by $A$ and $B$ respectively

Since $A$ and $B$ form a basis of $R^2$, then by **theorem 2.1**, each vector in $R^2$ can be written as a unique form of linear combination of $A$ and $B$. By definition $R^2$ is a direct sum of subspaces generated by $A$ and $B$ respectively.

Q.E.D.
