## Solving linear equations as an application of the dimension theorems

### System of linear equations

Let $K$ be a field. Let $A =(a_{i j})$, $i = 1, ..., m$ and $j = 1, ..., n$ be a matrix in $K$. Let $b_1 , ... ,b_m$ be elements of K. Equations like

$$
\begin{array}{c}
a_{1 1} x_1 + ... + a_{1 n} x_n &= b_1 \\
...  \\
a_{m 1} x_1 + ... + a_{m n} x_n &= b_m \\
\end{array}
$$ (\*)

where $x_1, ..., x_n \in K$, are called linear equations. (\*) is a **system of linear equations**.

$x_1, ..., x_n$ are called the unknowns. Hence $n$ is the number of unknowns.

$m$ is the number of equations.

$(a_{i j})$ is the matrix of coefficients.

### Homogeneous systems

The system is called **homogeneous** if $b_1, ..., b_m$ are equal to $0$.

$$
\begin{array}{c}
a_{1 1} x_1 + ... + a_{1 n} x_n &= 0 \\
...  \\
a_{m 1} x_1 + ... + a_{m n} x_n &= 0 \\
\end{array}
$$ (\*\*)

(\*\*) is called the **homogeneous system** associated with (\*).

The system (\*\*) always has a solution, which is $x_i = 0$ for $i$ in $1, ..., n$. This is called the **trivial solution**. A solution $(x_1, ..., x_n)$ to the system such that some $x_i \neq 0$ is called **non-trivial**.

The homogeneous system can be written as a linear combination of column vectors:

$$
\begin{equation}
\begin{aligned}
x_1 \begin{pmatrix} a_{1 1} \\ \vdots \\ a_{m 1} \end{pmatrix} + ... + x_n \begin{pmatrix} a_{1 n} \\ \vdots \\ a_{m n} \end{pmatrix} &= \begin{pmatrix} 0 \\ \vdots \\ 0 \end{pmatrix}
\end{aligned}
\end{equation}
$$

or let $A = (a_{i j})$, for $i$ in $1, ..., m$ and $n$ in $1, ..., n$, and let $A^j$ be the $j$-th column vector $\begin{pmatrix}a_{1 j} \\ \vdots \\ a_{m j}\end{pmatrix}$. Then above can be written as

$$
\begin{equation}
\begin{aligned}
x_1 A^1 + ... + x_n A^n &= \mathit{0}_m
\end{aligned}
\end{equation}
$$

Hence a non-trivial solution $X = (x_1, ..., x_n)$ is just an $n$-tuple/vector $X \neq \mathit{0}_n$, which gives a relation of **linear dependence** between the columns $A^1, ..., A^n$.

Now we can see that $A^j \in K^m$ for $j$ in $1, ..., n$, and $(x_1, ..., x_n) \in K^n$.

### Theorem 2.1

Let

$$
\begin{array}{c}
a_{1 1} x_1 + ... + a_{1 n} x_n &= 0 \\
... \\
a_{m 1} x_1 + ... + a_{m n} x_n &= 0
\end{array}
$$

be a homogenous system with $m$ linear equations and $n$ unknowns, with coefficient in a field $K$.

Assume $n > m$, then the system has a non-trivial solution in $K$.

#### Proof

From [theorem 3.1 in chapter 1](../1_vector_spaces/1_vector_spaces.ipynb#theorem_3.1):

> Let $V$ be a vector space over the field $K$. Let $\{v_1, ..., v_m\}$ be a basis of $V$ over $K$. Let $w_1, ..., w_n$ be elements of $V$, and assume that $n > m$. Then $w_1, ..., w_n$ are linearly dependent.

Since $\dim K^n = n$, and $\dim K^m = m$, and $n > m$, and since $A^j \in K^m$ for $j$ in $1, ..., n$, then we have $A^j \in K^m$ for $j$ in $1, ..., n$ are **linearly dependent**.

Hence there exist $x_1, ..., x_n \in K$ with some elements $x_j \neq 0$ for $j$ in $1, ..., n$ such that

$$
\begin{equation}
\begin{aligned}
x_1 A^1 + ... + x_n A^n &= \mathit{0}_m
\end{aligned}
\end{equation}
$$

Hence the homogeneous system has a non-trivial solution in $K$. Q.E.D.

Now we can consider the original system (\*).

Let $B$ be a column vector $B = \begin{pmatrix}b_1 \\ \vdots \\ b_m \end{pmatrix}$.

Then we may rewrite (\*) as a linear combination of column vectors

$$
\begin{equation}
\begin{aligned}
x_1 \begin{pmatrix} a_{1 1} \\ \vdots \\ a_{m 1} \end{pmatrix} + ... + x_n \begin{pmatrix} a_{1 n} \\ \vdots \\ a_{m n} \end{pmatrix} &= \begin{pmatrix} b_1 \\ \vdots \\ b_m \end{pmatrix}
\end{aligned}
\end{equation}
$$

or as

$$
\begin{equation}
\begin{aligned}
x_1 A^1 + ... + x_n A^n &= B
\end{aligned}
\end{equation}
$$

### Theorem 2.2. 

Assume that $m = n$ in the system (\*) above, and that the vectors $A_1, ..., A_n$ are linearly independent. Then the system (\*) has a solution in $K$, and this solution is unique.

#### Proof

Since $A_1, ..., A_n$ are linearly independent, and $m = n$, then $A_1, ..., A_n$ forms a basis of $K^n$. Hence any vector $B \in K^n$ can be written as a unique linear combination of $A_1, ..., A_n$. Hence there exists a unique $(x_1, ..., x_n)$ such that

$$
\begin{equation}
\begin{aligned}
x_1 A^1 + ... + x_n A^n &= B
\end{aligned}
\end{equation}
$$

Hence the solution to the system (\*) exists, and is unique. Q.E.D.
