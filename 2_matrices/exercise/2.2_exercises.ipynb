## 2. Let (\*\*) be a system of homogeneous linear equations in a field $K$, in $n$ unknowns. Show that the set of solutions $X = (x_1, ..., x_n)$ is a vector space over $K$.

### Proof

Let $V$ be the set of solutions to (\*\*). We know $V$ is a subset of $K^n$, which is a vector space.

Now we need to prove $V$ is a subspace of $K^n$.

Let $m$ be the number of equations of (\*\*), and let $A^1, ..., A^n$ be the column vectors of (\*\*).

Then we can write (\*\*) as a linear combination of column vectors:

$$
\begin{equation}
\begin{aligned}
x_1 A^1 + ... + x_n A^n &= \mathit{0}_{col}
\end{aligned}
\end{equation}
$$

(i): Let $v = (y_1, ..., y_n), w = (z_1, ..., z_n) \in V$. Then both $v$ and $w$ are solutions of (\*\*). Then

$$
\begin{equation}
\begin{aligned}
y_1 A^1 + ... + y_n A^n &= \mathit{0}_{col} \\
z_1 A^1 + ... + z_n A^n &= \mathit{0}_{col}
\end{aligned}
\end{equation}
$$

By summing up the two linear combinations, we have

$$
\begin{equation}
\begin{aligned}
(y_1 + z_1) A^1 + ... (y_n + z_n) A^n &= \mathit{0}_{col} + \mathit{0}_{col} \\
&= \mathit{0}
\end{aligned}
\end{equation}
$$

Hence $v + w = (y_1 + z_1, ..., y_n + z_n)$ is also a solution of (\*\*). Hence $(v + w) \in V$.

(ii): Let $K$ be the field under (\*\*). Let $v = (y_1, ..., y_n) \in V$, and $c \in K$ Then

$$
\begin{equation}
\begin{aligned}
y_1 A^1 + ... + y_n A^n &= \mathit{0}_{col}
\end{aligned}
\end{equation}
$$

By multplying the linear combinations by $c$, we have

$$
\begin{equation}
\begin{aligned}
c (y_1 A^1 + ... + y_n A^n) &= c \mathit{0}_{col} \\
(c y_1) A^1 + ... + (c y_n) A^n &= \mathit{0}_{col}
\end{aligned}
\end{equation}
$$

Hence $c v = (c y_1, ..., c y_n)$ is also a solution to (\*\*). Hence $c v \in V$.

(iii): The zero vector in $K^n$ is
$$
\begin{equation}
\begin{aligned}
\mathit{0} &= (\underbrace{0, ..., 0}_n)
\end{aligned}
\end{equation}
$$

It is obivous that $\mathit{0}$ is a solution to the homogeneous system (\*\*).

Hence $\mathit{0} \in V$.

Q.E.D.

## 3. Let $A_1, ..., A_n$ be column vectors of size $m$. Assume that they have coefficients in $R$, and that they are linearly independent over $R$. Show that they are linearly independent over $C$.

### Proof

Since $A_1, ..., A_n$ are column vectors of size $m$ that are linearly independent in $R$, there is no non-trival solution to the system of equations that is

$$
x_1 A^1 + ... x_n A^n = \mathit{0}_{col(R^m)}
$$

for $x_1, ..., x_n \in R$.

Let $c_1, ..., c_n \in C$ where $c_j = y_j + z_j i$ for $y_j, z_j \in R$ and $j$ in $1, ..., n$ such that

$$
c_1 A^1 + ... + c_n A^n = \mathit{0}_{col(C^m)}
$$

Then

$$
\begin{array}{c}
(y_1 + z_1 i) A^1 + ... + (y_n + z_n i) A^n &= \mathit{0}_{col(C^m)} \\
(y_1 A^1 + ... + y_n A^n) + i (z_1 A^1 + ... + z_n A^n) &= \mathit{0}_{col(R^m)} + i \mathit{0}_{col(R^m)}
\end{array}
$$

Hence

$$
\begin{array}{c}
y_1 A^1 + ... + y_n A^n &= \mathit{0}_{col(R^m)} \\
z_1 A^1 + ... + z_n A^n &= \mathit{0}_{col(R^m)}
\end{array}
$$

And since there is no non-trival solution to

$$
x_1 A^1 + ... x_n A^n = \mathit{0}_{col(R^m)}
$$

Then the only solution to $y_1, ..., y_n$ and $z_1, ..., z_n$ is

$$
\begin{equation}
\begin{aligned}
\mathit{0}_{R^n} &= (\underbrace{0, ..., 0}_n)
\end{aligned}
\end{equation}
$$

Hence the only solution to

$$
c_1 A^1 + ... + c_n A^n = \mathit{0}_{col(C^m)}
$$

is

$$
\begin{equation}
\begin{aligned}
\mathit{0}_{C^n} &= (\underbrace{0 + 0 i, ..., 0 + 0 i}_n)
\end{aligned}
\end{equation}
$$

Hence $A^1, ..., A^n$ are linearly independent over $C$. Q.E.D.

## 4. Let (\*\*) be a system of homogeneous linear equations with coefficients in $R$. If this system has a non-trivial solution in $C$, show that it has a non-trivial solution in $R$.

### Proof

Let $m$ be the number of equations, and $n$ be the number of unknowns.

Let $A^1, ..., A^n$ be the $m$-size column vectors of (\*\*).

As (\*\*) has a non-trival solution in $C$, there exist $c_1, ..., c_n \in C$ where $c_j = (y_j + z_j i)$ where there are some $(y_j + z_j i) \neq (0 + 0 i)$ for $y_j, z_j \in R$ and $j$ in $1, ..., n$ such that

$$
\begin{array}{c}
c_1 A^1 + ... + c_n A^n &= \mathit{0}_{col(C^m)} \\
(y_1 + z_1 i) A^1 + ... + (y_n + z_n i) A^n &= \mathit{0}_{col(R^m)} + i \mathit{0}_{col(R^m)} \\
(y_1 A^1 + ... + y_n A^n) + i (z_1 A^1 + ... + z_n A^n) &= \mathit{0}_{col(R^m)} + i \mathit{0}_{col(R^m)}
\end{array}
$$

Hence there exist some $y_j \neq 0$ or $z_j \neq 0$ such that
$$
\begin{array}{c}
y_1 A^1 + ... + y_n A^n &= \mathit{0}_{col(R^m)} & \text{or} \\
z_1 A^1 + ... + z_n A^n &= \mathit{0}_{col(R^m)}
\end{array}
$$

Hence (\*\*) has a non-trivial solution in $R$. Q.E.D.
