We are defining the notion of a **mapping**, which generalize the notion of a function. 

Among the mappings, the **linear** mappings are the most important.
- A good deal of mathematics is devoted to reduce question concerning arbitrary mappings to linear ones
- They are interesting by themselves
- Many mappings can be approximated with a linear one, which is done in calculus of several variables

# Definition: Mappings

Let $S, S^{\prime}$ be two sets.

A **mapping** from $S$ to $S^\prime$ is an association which to every element of $S$ associate an element of $S^\prime$.

For a mapping $F$ from $S$ to $S^\prime$, we can write $F: S \to S^\prime$.

## Functions as a special kind of mappings

A function is a mapping from a set into the set of numbers, i.e. into $R$, or $C$, or a field $K$. (Note: only Serge Lang makes such definition of functions.)

## Value/Image

If $T: S \to S^\prime$ is a mapping, and if $u \in S$, then we denote the element of $S^\prime$ associate with $u$ as $T(u)$, or $Tu$.

We call $T(u)$ the **value** of $T$ at $u$, or also the **image** of $u$ under $T$.

$T(u)$ is read $T$ of $u$.

For $W \subseteq S$, then $\{T(w) | w \in W\}$ is called the image of $W$ under $T$. We denote $T(W) = \{T(w) | w \in W\}$.

For $F: S \to S^\prime$, if $x \in S$, we often write

$$
x \mapsto F(x)
$$

with a special arrow $\mapsto$ to denote the image of $x$ under $F$. Thus for example we would speak of the map $F$ such that $F(x) = x^2$ as the map $x \mapsto x^2$.

### Example 6

Let $F: R^2 \to R^2$ be the mapping given by

$$
F(x, y) = (2 x, 2 y)
$$

Describe the image under $F$ of the points lying on the circle $x^2 + y^2 = 1$.

#### Solution

Let $(x, y)$ be a point on the circle of radius $1$.

Let $u = 2 x$, $v = 2 y$. Then $u, v$ satisfy the relation

$$
(\frac{u}{2})^2 + (\frac{v}{2})^2 = 1
$$

Thus

$$
\begin{array}{c}
\frac{u^2}{4} + \frac{v^2}{4} &= 1 \\
u^2 + v^2 &= 2^2
\end{array}
$$

Hence $(u, v)$ is a point on the circle of radius $2$. Therefore the image under $F$ of the circle of radius $1$ is a subset of the circle of radius $2$.

Conversely, given a point $(u, v)$ such that

$$
u^2 + v^2 = 4
$$

Let $x = \frac{u}{2}$ and $y = \frac{v}{2}$. Then we have

$$
x^2 + y^2 = (\frac{u}{2})^2 + (\frac{v}{2})^2 = \frac{u^2 + v^2}{4} = \frac{4}{4} = 1
$$

Hence $(x, y)$ is a point on the circle of radius $1$. Hence every point on the circle of radius $2$ is the image of some point on the circle of radius $1$.

We conclude finally that the image of the circle of radius $1$ under $F$ is precisely the circle of radius $2$.

**Note:** In general, let $S, S^\prime$ be two sets. To prove $S = S^\prime$, one frequently proves that $S \subseteq S^\prime$ and $S^\prime \subseteq S$. This is what we did in the preceding argument.

### Example 7

Let $S$ be a set and let $V$ be a vector space over the field $K$.

Let $F, G$ be mappings of $S$ into $V$.

We can define their sum $F + G$ as the map whose value at an element $t$ of $S$ is $F(t) + G(t)$.

We also define the product of $F$ by a $c \in K$ to be the map whose value at $t \in S$ is $cF(t)$.

Since $V$ is a vector space, it is easy to verify that **VS 1** through **VS 2** are satisfied for the set of all mappings of $S$ into $V$.

## Example 8

Let $S$ be a set.

Let $F: S \to K^n$ be a mapping. For each element $t \in S$, the value of $F$ at $t$ is a vector $F(t)$. The coordinates of $F(t)$ depend on $t$. Hence there are functions $f_1, ..., f_n$ of $S$ into $K$ such that

$$
F(t) = (f_1(t), ..., f_n(t))
$$

$f_1, ..., f_n$ are called the **coordinate functions** of $F$.

Since $K^n$ is a vector space, from the conclusion in **example 7**, let there be $F, G: S \to K^n$, and

$$
\begin{array}{c}
F(t) &= (f_1(t), ..., f_n(t)) \\
G(t) &= (g_1(t), ..., g_n(t)) \\
\end{array}
$$

$$
(F + G)(t) = F(t) + G(t) = (f_1(t) + g_1(t), ..., f_n(t) + g_n(t))
$$

Hence the coordinate functions of $F + G$ are $f_1 + g_1, ..., f_n + g_n$.

And for any $c \in K$

$$
(c F)(t) = c F(t) = (c f_1(t), ..., c f_n(t))
$$

Hence the coordinate functions of $c F$ are $c f_1, ..., c f_n$.

### Curve in $R^n$

For instance, let $J$ be an interval of real numbers, then

$$
F: J \to R^n
$$

is also called a **(parametric) curve** in $n$-space.

## Mapping composition

Let $U, V, W$ be sets. Let $F: U \to V$ and $G: V \to W$ be mappings. Then we can form the composite mapping from $U$ into $W$, denoted by $G \circ F$. It is by definition the mapping defined by

$$
\begin{equation}
\begin{aligned}
(G \circ F)(t) &= G(F(t)) & \text{for all $t \in U$}
\end{aligned}
\end{equation}
$$

### Associativity of mapping compositions

Let $U, V, W, S$ be sets. Let

$$
\begin{array}{c}
F: U \to V \\
G: V \to W \\
H: W \to S
\end{array}
$$

be mappings. Show that

$$
H \circ (G \circ F) = (H \circ G) \circ F
$$

#### Proof

By definition

$$
H \circ (G \circ F)(t) = H((G \circ F)(t)) = H(G(F(t)))
$$

and

$$
((H \circ G) \circ F)(t) = (H \circ G)(F(t)) = H(G(F(t)))
$$

Hence

$$
H \circ (G \circ F) = (H \circ G) \circ F
$$

Q.E.D.

## Injective mappings (One-to-one)

Let

$$
f: S \to S^\prime
$$

be a map. We say that $f$ is **injective** if whenever $x, y \in S$ and $x \neq y$, then $f(x) \neq f(y)$.

Put it another way, we can say

$$
f(x) = f(y) \implies x = y
$$

For example, $f: R \to R$ such that $f(x) = x^2$ is not injective, as $f(1) = f(-1) = 1$.

However, $f: R \to R$ such that $f(x) = x + 1$ is injective.

## Surjective mappings (Onto)

Let

$$
f: S \to S^\prime
$$

be a map. We say that $f$ is **surjective** if the image of $f$ is all of $S^\prime$.

For example, $f: R \to R$ such that $f(x) = x^2$ is not surjective, as its image is $\{y \geq 0 | y \in R\} \subsetneq R$.

However,  $f: R \to R$ such that $f(x) = x^3$ is surjective, as for each number in $R$ its cube root always exists.

A map is called **bijective** if it is both injective and surjective.

### Being injective/surjective is related to domain and image

Let $R^+$ be the set of real numbers $\geq 0$. As a matter of convention, we agree to distinguish between the maps

$$
R \to R
$$

and

$$
R^+ \to R^+
$$

given by the same formula

$$
x \mapsto x^2
$$

For $x \mapsto x^2$ as a map from $R$ into $R$, the mapping is not injective neither surjective.

But for $x \mapsto x^2$ as a map from $R^+$ into $R^+$, the mapping is bijective.

In general, when dealing with a map $f: s \to s^\prime$, we must therefore always specify the sets $S$ and $S\prime$, to be able to say that $f$ is injective, or surjective, or neither. To have a completely accurate notation, we should write

$$
f_{S, S^\prime}
$$

but this becomes too clumsy, and we prefer to use the context to make our meaning clear.

## Identity mapping

If $S$ is any set, the **identity mapping** $I_s$ is defined to be the map that
$$
\begin{equation}
\begin{aligned}
I_S(x) &= x & \text{for all $x \in S$}
\end{aligned}
\end{equation}
$$

The identity mapping is bijective.

## Inverse mapping

Let $F: S \to S^\prime$. We say $F$ has an **inverse** if there exists a mapping $G: S^\prime \to S$ such that

$G \circ F = I_S$ and $F \circ G = I_{S^\prime}$.

[If an inverse mapping exits for a mapping, then it is unique.](https://math.stackexchange.com/a/2187768)


###  Example 11

Let $S$ be the set of all real numbers $> 0$.

Let $f: S \to S$ be the map such that $f(x) = x^2$.

Then $f$ has an inverse mapping, namely $g: S \to S$ such that $g(x) = \sqrt{x}$.

## Translations

Let $V$ be a vector space, and let $u$ be a fixed element of $V$.

Let $T_u: V \to V$ be the map such that $T_u(v) = u + v$.

If $S \subseteq V$, then $T_u(S)$ is called the **translation of $S$ by $u$**.

### Translations and compositions

If $u_1, u_2 \in V$, then $T_{u_1 + u_2} = T_{u_1} \circ T_{u_2}$

#### Proof

For some $v \in V$, we have

$$
T_{u_1 + u_2}(v) = v + (u_1 + u_2) = (v + u_1) + u_2 = T_{u_1}(T_{u_2}(v)) = (T_{u_1} \circ T_{u_2})(v)
$$

Thus

$$
T_{u_1 + u_2} = T_{u_1} \circ T_{u_2}
$$

Q.E.D.

### Translations and inverses

If $u \in V$, then $T_u: V \to V$ has an inverse mapping which is nothing but the translation $T_{-u}$.

#### Proof

For some $v \in V$, we have

$$
T_u \circ T_{- u} (v) = T_u(T_{- u}(v)) = T_u(v + (- u)) = (v + (- u)) + u = v
$$

Hence $T_u \circ T_{- u} = I$.

And

$$
T_{- u} \circ T_u (v) = T_{- u}(T_u(v)) = T_{- u}(v + u) = (v + u) + (- v) = v
$$

Hence $T_{- u} \circ T_u = I$.

Hence $T_{-u}$ is an inverse mapping of $T_u$. And [the inverse mapping of a mapping is unique](#Inverse-mapping). Q.E.D.

## Bijectiveness and inverse existence

Let $f: S \to S^\prime$ be a mapping.

$f$ is bijective if and only if it has an inverse mapping.

### Proof

Prove the forward statement: If $f$ has an inverse mapping then $f$ is bijective.

Let $g: S^\prime \to S$ be the inverse mapping of $f$.

Let $x, y \in S$ such that $f(x) = f(y)$. Since $f \circ g = g \circ f = I$, then we have

$$
x = g(f(x)) = g(f(y)) = y
$$

Hence $f$ is injective.

Let $z \in S^\prime$. Then we have

$$
f(g(z)) = z
$$

Hence for every $z \in S^\prime$ there exists $x = g(z)$ such that $f(x) = z$. Hence $f$ is surjective.

Prove the backward statement: If $f$ is bijective then $f$ has an inverse mapping.

Let $S^{\prime\prime}$ be the image of $f$. Since $f$ is injective, there exists a mapping $g: S^{\prime\prime} \to S$ such that if $f(x) = y$ then $g(y) = x$ for $x \in S$ and $y \in S^{\prime\prime}$. Hence $g(f(x)) = x$ and $f(g(y)) = y$.

And since $f$ is surjective, $S^{\prime\prime} = S^\prime$.

Hence there exists a mapping $g: S^\prime \to S$ such that

$$
g(f(x)) = x
$$

and

$$
f(g(y)) = y
$$

Hence $f$ has an inverse $g$.

Q.E.D.
