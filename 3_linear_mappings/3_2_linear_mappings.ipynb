## Definition: Linear Mappings

Let $V, V^\prime$ be the vector spaces over the field $K$.

A **linear mapping** $F: V \to V^\prime$ is a mapping which satisfies the following two properties
- **LM 1**: For any elements $u, v \in V$ we have

$$
F(u + v) = F(u) + F(v)
$$
- **LM 2**: For all $c \in K$ and $v \in V$ we have

$$
F(c v) = c F(v)
$$

If we wish to specify the field $K$, we also say that $F$ is $K$-linear. Since we usually deal with a fixed field $K$, we omit the prefix $K$, and say simply that $F$ is **linear**.

### Example 1

Let $V$ be a finite dimensional space over $K$, and let $\{v_1, ..., v_n\}$ be a basis of $V$.

We define map $F: V \to F^n$ by associating to each element $v \in V$ its coordinate $X = (x_1, ..., x_n)$ with respect to the basis. 

Thus if

$$
v = x_1 v_1 + ... + x_n v_n
$$

we let

$$
F(v) = (x_1, ..., x_n)
$$

Now show that $F$ is a linear map.

#### Proof

**LM 1**

Let $u, w \in V$. Since $\{v_1, ..., v_n\}$ be a basis of $V$, then we can write $u, w$ as

$$
\begin{array}{c}
u &= a_1 v_1 + ... + a_n v_n \\
w &= b_1 v_1 + ... + b_n v_n
\end{array}
$$

For $a_1, ..., a_n$ and $b_1, ..., b_n \in F$.

Thus

$$
u + w = (a_1 + b_1) v_1 + ... + (a_n + b_n) v_n
$$

Then

$$
F(u + w) = F((a_1 + b_1) v_1 + ... + (a_n + b_n) v_n) = (a_1 + b_1, ...,  a_n + b_n) = (a_1, ..., a_n) + (b_1, ..., b_n) = F(u) + F(w)
$$

**LM 2**

Let $c \in K$. Then we have

$$
c u = c (a_1 v_1 + ... + a_n v_n) = (c a_1) v_1 + ... + (c a_n) v_n
$$

Thus

$$
F(c u) = F((c a_1) v_1 + ... + (c a_n) v_n) = (c a_1, ..., c a_n) = c (a_1, ..., a_n) = c F(u)
$$

Q.E.D.

### Example 2

Let $V = R^3$ be the vector space (over $R$) of vectors in $3$-space. Let $V^\prime = R^2$ be the vector space of vectors in $2$-space. We can define a mapping $F: R^3 \to R^2$ by the projection, namely

$$
F(x, y, z) = F(x, y)
$$

Proof $F$ is linear.

#### Proof

Let $(x_1, y_1, z_1), (x_2, y_2, z_2) \in R^3$.

**LM 1**

$$
F((x_1, y_1, z_1) + (x_2, y_2, z_2)) = F(x_1 + x_2, y_1 + y_2, z_1 + z_2) = (x_1 + x_2, y_1 + y_2) = (x_1, y_1) + (x_2, y_2) = F(x_1, y_1, z_1) + F(x_2, y_2, z_2)
$$

**LM 2**

Let $c \in R$. We have

$$
F(c(x_1, y_1, z_1)) = F(c x_1, c y_1, c z_1)= (c x_1, c y_1)= c (x_1, y_1) = c F(x_1, y_1, z_1)
$$

Q.E.D.

### Example 3

Let $A =(1, 2, -1)$. We can define a mapping $L = L_A: R^3 \to R$ by the association $X \mapsto X \cdot A$, i.e. $L(X) = X \cdot A$ for any vector $X$ in $R^3$.

The fact that $L$ is linear summarizes two known properties of the scalar product, namely, for any vectors $X, Y \in R^3$ we have

$$
(X + Y) \cdot A = X \cdot A + Y \cdot A \\
(c X) \cdot A = c (X \cdot A)
$$

More generally, let $K$ be a field, and A a fixed vector in $K_n$. We have a linear map (i.e. $K$-linear map) $L_A: K^n \to K$ such that $L_A(X) = X \cdot A$ for all $X \in K^n$.

#### Generalization to matrix multiplications

Let $A$ be an $m \times n$ matrix in a field $K$. We obtain a linear map $L_A: K^n \to K$ such that

$$
L_A(X) = A X
$$

for every column vector $X$ in $K^n$. The linearity follows from properties of matrix multiplications.

### Identity mapping is linear

Let $V, V^\prime$ be vector spaces over field $K$. Let $I: V \to V^\prime$ be the identity mapping. Let $u, v \in V$. Let $c \in K$. Then

$$
I(u + v) = u + v = I(u) + I(v) \\
I(c u) = c u = c I(i)
$$

Hence an identity mapping from a vector space into another vector space over the same field is linear.

### Zero mapping is linear

Let $V, V^\prime$ be vector spaces over field $K$. Let $u \in V$ Let $\mathit{0}: V \to V^\prime$ be the **zero** mapping such that. 

$$
\begin{equation}
\begin{aligned}
\underbrace{\mathit{0}}_{\text{mapping}}(u) &= \underbrace{\mathit{0}}_{\text{vector in $V^\prime$}} \\
\end{aligned}
\end{equation}
$$

Let $u, v \in V$. Let $c \in K$. Let $c \in K$. Then

$$
\mathit{0}(u + v) = \mathit{0} = \mathit{0}(u) + \mathit{0}(v) \\
\mathit{0}(c u) = \mathit{0} = c \mathit{0} = c \mathit{0}(v)
$$

### Linear mapping at zero-vector $\mathit{0}$

Show that 

Let $L: V \to W$ be a linear map. Then $L(\mathit{0})=\mathit{0}$.

#### Proof

Let $u \in V$. By [VS 3](../../1_vector_spaces/1_vector_spaces.ipynb#Definition), there exists a $- u$ such that $u + (- u) = \mathit{0}$.

Hence

$$
\begin{equation}
\begin{aligned}
L(\underbrace{\mathit{0}}_{\in V}) &= L(u + (-u)) \\
&= L(u) + L(-u) &\text{LM 1} \\
&= L(u) - L(u) &\text{LM 2} \\
&= \underbrace{\mathit{0}}_{\in W}
\end{aligned}
\end{equation}
$$

Q.E.D.

And as $P \to Q \implies \neg Q \to \neg P$, we have

If $F: V \to W$ is a mapping and $F(\mathit{0}) \neq \mathit{0}$ then $F$ is not linear.

### Example 6: The space of linear maps

Let $V, V^\prime$ be two vector spaces over the field $K$.

We consider the set of **all** linear mappings from $V$ into $V^\prime$, and denote this set by $\mathscr{L}(V, V^\prime)$, or simply $\mathscr{L}$ if the reference to $V, V^\prime$ is clear.

We shall define the addition of linear mappings and their multiplication by numbers in such a way as to make $\mathscr{L}$ into a vector space.

#### Sum

Let $T: V \to V^\prime$ and $F: V \to V^\prime$ be two linear mappings. We define their sum $T + F$ to be the map whose value at an element $u \in V$ is $T(u) + F(u)$. Thus we may write

$$
(T + F)(u) = T(u) + F(u)
$$

Then $T + F$ is a linear map because:
- **LM 1**: Let $u, v \in V$, then $(T + F)(u + v) = T(u + v) + F(u + v) = T(u) + T(v) + F(u) + F(v) = (T(u) + F(u)) + (T(v) + F(v)) = (T + F)(u) + (T + F)(v)$.
- **LM 2**: Let $c \in K$, then $(T + F)(c u) = T(c u) + F(c u) = c T(u) + c F(u) = c (T(u) + F(u)) = c((T + F)(u))$

#### Scalar multiplication

If $a \in K$, and $T: V \to V^\prime$ is a linear map, we define a map $a T: V \to V^\prime$ as

$$
(a T)(u) = a T(u)
$$

#### $\mathscr{L}$ as a vector space

First, for any $T \in \mathscr{L}(V, V^\prime)$, we define $- T = (- 1) T$.

Then we define the **zero-map** $\mathit{0}: V \to V^\prime$ as

$$
\begin{equation}
\begin{aligned}
\underbrace{\mathit{0}}_{\text{zero-map}}(v) &= \underbrace{\mathit{0}}_{\text{zero-vector in $V^\prime$}} & \text{for all $v \in V$}
\end{aligned}
\end{equation}
$$

We now show that $\mathscr{L}(V, V^\prime)$ is a vector space.

##### Proof

**VS 1** Additive associativity: For any $T, F, G \in \mathscr{L}(V, V^\prime)$, and for any $v \in V$, since $V^\prime$ is a vector space, we have $((T + G) + F)(v) = (T + G)(v) + F(v) = T(v) + (G(v) + F(v)) = T(v) + (G + F)(v) = (T + (G + F))(v)$. Hence $(T + G) + F = T + (G + F)$.

**VS 2** Additive identity: For any $T \in \mathscr{L}(V, V^\prime)$, and for any $v \in V$, for the zero-map $\mathit{0}$, we have

$$
\begin{equation}
\begin{aligned}
(T + \underbrace{\mathit{0}}_{\text{zero-map}})(v) &= T(v) + \underbrace{\mathit{0}}_{\text{zero-map}}(v) \\
&= T(v) + \underbrace{\mathit{0}}_{\text{zero-vector in $V^\prime$}} \\
&= T(v) \\
&= \underbrace{\mathit{0}}_{\text{zero-vector in $V^\prime$}} + T(v) \\
&= (\underbrace{\mathit{0}}_{\text{zero-map}} + T)(v)
\end{aligned}
\end{equation}
$$

Hence

$$
T + \underbrace{\mathit{0}}_{\text{zero-map}} = T = \underbrace{\mathit{0}}_{\text{zero-map}} + T
$$

**VS 3** Additive inverse: For any $T \in \mathscr{L}(V, V^\prime)$, there exists a $- T$ such that for any $v \in V$

$$
(T + (- T)))(v) = (T + (-1) T)(v) = T(v) + (-1) T(v) = \underbrace{\mathit{0}}_{\text{zero-vector in $V^\prime$}}
$$

**VS 4** Commutativity: For any $T, F \in \mathscr{L}(V, V^\prime)$, and for any $v \in V$, since $V^\prime$ is a vector space, we have $(T + F)(v) = T(v) + F(v) = F(v) + T(v) = (F + T)(v)$. Hence $T + F = F + T$.

**VS 5** Distributivity of scalar multiplication with respect to vector addition: For any $T F \in \mathscr{L}(V, V^\prime)$, for any $v \in V$, and for any $c \in K$, since $V^\prime$ is a vector space, we have $(c (T + F))(v) = c (T(v) + F(v)) = c T(v) + c F(v)$. Hence $c (T + F) = c T + c F$.

**VS 6** Distributivity of scalar multiplication with respect to field addition: For any $T \in \mathscr{L}(V, V^\prime)$, for any $v \in V$, and for any $a, b \in K$, since $V^\prime$ is a vector space, we have $((a + b)F)(v) = (a + b)(F(v)) = a F(v) + b F(v)$. Hence $(a + b) F = a F + b F$.

**VS 7.** Compatibility of scalar multiplication with field multiplication: For any $T \in \mathscr{L}(V, V^\prime)$, for any $v \in V$, and for any $a, b \in K$, since $V^\prime$ is a vector space, we have $((a b) F)(v) = (a b) F(v) = a (b F(v)) = a ((b f)(v))$. Hence $(a b) F = a (b F)$.

**VS 8** Identity element of scalar multiplication: For any $T \in \mathscr{L}(V, V^\prime)$, for any $v \in V$, we have $(1 F)(v) =  1 \cdot F(v) = F(v)$. Hence $1 F = F$.

Hence $\mathscr{L}(V, V^\prime)$ is a vector space. Q.E.D.

### Example 7

Let $V$ be the vector space of real valued functions of a real variable which have derivatives of all order.

Let $\operatorname{D}$ be the derivative. Then $\operatorname{D}: V \to V$ is a linear map. This is merely a brief way of summarizing known properties of the derivative, namely

$$
D(f + g) = D f + D g
$$

and

$$
D(c f) = c D f
$$

for any $f, g \in V$ and for any $c \in R$.

Let $I: V \to V$ be the identity map. Then we have

$$
(D + I)f = D f + f
$$

for any $f \in V$.

#### Example: $f(x) = e^x$

For $f \in V$ where $f(x) = e^x$, we have

$$
((D + I) f)(x) = (D + I) e^x = D e^x + e^x = e^x + e^x = 2 e^x
$$

#### Example: $f(x) = \sin x$

For $f \in V$ where $f(x) = \sin x$, we have

$$
((D + I) f)(x) = (D + I) \sin x = D \sin x + \sin x = \cos x + \sin x
$$

### Linear mappings and linear combinations as variables

Let $S, S^\prime$ be vector spaces over $K$. Let $T: S \to S^\prime$ be a linear mapping. Let $u_1, u_2, ..., u_n \in S$, and let $x_1, x_2, ..., x_n \in K$. Show that

$$
T(x_1 u_1 + x_2 u_2  + ... + x_n u_n) = x_1 T(u_1) + x_2 T(u_2) + ... + x_n T(u_n)
$$

#### Proof

$$
\begin{equation}
\begin{aligned}
T(x_1 u_1 + x_2 u_2  + ... + x_n u_n) &= T(x_1 u_1 + (x_2 u_2, ..., x_n u_n)) \\
&= T(x_1 u_1) + T(x_2 u_2, ..., x_n u_n) & \text{LM 1 of $T$} \\
&... \\
&= T(x_1 u_1) + T(x_2 u_2) + ... + T(x_n u_n) & \text{LM 1 of $T$} \\
&= x_1 T(u_1) + x_2 T(u_2) + ... + x_n T(u_n) & \text{LM 2 of $T$}
\end{aligned}
\end{equation}
$$

Q.E.D.

### Theorem 2.1: Linear mappings and basis

Let $V$ and $W$ be vector spaces over $K$.

Let $\{v_1, ... ,v_n\}$ be a basis of $V$, and let $w_1, ..., w_n$ be arbitrary elements of $W$.

Then there exists a *unique* linear mapping $T: V \to W$ such that

$$
T(v_1) = w_1, ..., T(v_n) = w_n
$$

Let $K$ be the field under $V$. If $x_1, ..., x_n \in K$,

$$
T(x_1 v_1 + ... + x_n v_n) = x_1 w_1 + ... x_1 w_n
$$

#### Proof

First, we shall proof such a linear map $T$ exists.

Let $v \in V$. Since $\{v_1, ..., v_n\}$ is a set of basis of $V$, then there exist *unique* numbers $x_1, ..., x_n \in K$ such that

$$
v = x_1 v_1 + ... + x_n v_n
$$

Let

$$
T(v) = x_1 w_1 + ... + x_n w_n
$$

Now we can show $T$ is linear.

Let $v^\prime \in V$, and we can write $v^\prime = y_1 v_1 + ... + y_n v_n$ for some unique $y_1, ..., y_n \in K$. Then

$$
v + v^\prime = (x_1 + y_1) v_1 + ... + (x_n + y_n) v_n
$$

Then by definition, we have

$$
\begin{equation}
\begin{aligned}
T(v + v^\prime) &= (x_1 + y_1) w_1 + ... + (x_n + y_n) w_n \\
&= (x_1 w_1 + ... + x_n w_n) + (y_1 w_1 + ... + y_n w_n) \\
&= T(v) + T(v^\prime)
\end{aligned}
\end{equation}
$$

Hence **LM 1** is satisfied.

Let $c \in K$. Then $c v = c x_1 v_1 + ... c x_n v_n$

Hence

$$
\begin{equation}
\begin{aligned}
T(c v) &= c x_1 + ... c x_n \\
&= c (x_1 + ... x_n) \\
&= c T(v)
\end{aligned}
\end{equation}
$$

Hence **LM 2** is satisfied.

Hence $T$ is a linear mapping. Hence such a linear map exists.

Let $F: V \to W$ be a linear mapping where $F(v_i) = w_i$. Since for any $v \in V$, there exists unique $x_1, ..., x_n \in K$ such that $v = x_1 v_1 + ... x_n v_n$, then

$$
F(v) = F(x_1 v_1 + ... x_n v_n) = x_1 F(v_1) + ... + x_n F(v_n) = x_1 w_1 + ... + x_n w_n = T(v)
$$

Hence $F = T$. Hence such a linear mapping T is unique and

$$
T(v) = x_1 w_1 + ... + x_n w_n
$$


Q.E.D.

This theorem will be useful when exploring the association between linear mappings and matrices.

### Lemma 2.2: Linear mappings equivalence and basis

Let $V, W$ be vector spaces. Let $f, g \in \mathscr{L}(V, W)$.

Let $\{v_1, ..., v_n\}$ be a basis of $V$.

If $f(v_1) = g(v_1), ..., f(v_n) = g(v_n)$ then $f = g$.

#### Proof

Let $v$ be any vector in $V$. Since $\{v_1, ..., v_n\}$ be a basis of $V$, there exists some unique numbers $x_1, ..., x_n$ such that $v = x_1 v_1 + ... + x_n v_n$.

Then we can write $f(v), g(v)$ as

$$
f(v) = f(x_1 v_1 + ... + x_n v_n)
g(v) = g(x_1 v_1 + ... + x_n v_n)
$$

And since $f$ and $g$ are linear maps, we have

$$
f(x_1 v_1 + ... + x_n v_n) = x_1 f(v_1) + ... + x_n f(v_n)
g(x_1 v_1 + ... + x_n v_n) = x_1 g(v_1) + ... + x_n g(v_n)
$$

And since $f(v_1) = g(v_1), ..., f(v_n) = g(v_n)$, we have

$$
x_1 f(v_1) + ... + x_n f(v_n) = x_1 g(v_1) + ... + x_n g(v_n)
$$

Hence

$$
f(v) = g(v)
$$

Thus

$$
f = g
$$

Q.E.D.
