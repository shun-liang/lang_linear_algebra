## Definition: Kernel

Let $V, W$ be vector spaces over $K$, and let $F: V \to W$ be a linear map. We define the **kernel** of $F$ to be the set of elements $v \in V$ such that $F(v) = \mathit{0}$.

## Kernel as a subspace

The kernel of a linear map $F: V \to W$ is a subspace of $V$.

### Proof

(i) Let $v, u$ be elements of the kernel of $F$, Hence $F(v) = F(u) = \mathit{0}$. Since $F$ is a linear map, we have $F(v + u) = F(v) + F(u) = \mathit{0}$. Thus $v + u$ is also in the kernel of $F$.

(ii) Let $v$ be an element in the kernel of $F$, and let $c$ be an element in $K$. Since $F$ is a linear map, we have $F(c v) = c F(v) = c \mathit{0} = \mathit{0}$. Hence $c v$ is also in the kernel of $F$.

(iii) From the conclusion of [exercise 3.2](exercise/3.2_exercises.ipynb#2.-Let-$T:-V-\to-W$-be-a-linear-map-from-one-vector-space-into-another.-Show-that-$T(\mathit{0})-=-\mathit{0}$.), since $F$ is a linear map, we know $F(\mathit{0}) = \mathit{0}$.

Q.E.D.

## Kernel and injectivity

Let $F: V \to W$ be a linear map. We contend that the following two are equivalent:

1. The kernel of $F$ is equal to $\{\mathit{0}\}$.
2. If $v, w$ are elements of $V$ such that $F(v) = F(w)$, then $v = w$. In other words, $F$ is injective.

### Proof

#### If 1 then 2

Assume the kernel of $F$ is equal to $\{\mathit{0}\}$.

Let $v, w$ be elements in $V$ such that $F(v) = F(w)$. Then $F(v - w) = F(v) - F(w) = \mathit{0}$. Hence $v - w$ is an element in the kernel of $F$. Thus $v - w = \mathit{0}$. Hence $v = w$, thus $F$ is injective.

#### If 2 then 1

Assume $F$ is injective.

Let $v, w$ be elements of the kernel of $V$. Thus $F(v) = F(w) = \mathit{0}$. Since $F$ is injective, we have $v = w$.

And by the conclusion of [exercise 3.2](exercise/3.2_exercises.ipynb#2.-Let-$T:-V-\to-W$-be-a-linear-map-from-one-vector-space-into-another.-Show-that-$T(\mathit{0})-=-\mathit{0}$.), since $F$ is a linear map, we know $F(\mathit{0}) = \mathit{0}$.

Hence for every $v$ in the kernel of $V$, we have $v = \mathit{0}$. Hence the kernel of $F$ is equal to $\{\mathit{0}\}$.

Q.E.D.

## Theorem 3.1.

Let $F: V \to W$ be a linear map whose kernel is $\{O\}$. If $v_l , ... ,v_n$ are linearly independent elements of $V$, then $F(v_l), ... ,F(v_n)$ are linearly independent elements of $W$.

### Proof

Assume $F(v_1), ..., F(v_n)$ are linearly dependent in $W$. Hence there exists some numbers $a_1, ..., a_n$ which are not all $\neq 0$ such that 

$$
a_1 F(v_1) + ... + a_n F(v_n) = \mathit{0}
$$

Since $F$ is a linear map, we have

$$
F(a_1 v_1 + ... + a_n v_n) = a_1 F(v_1) + ... + a_n F(v_n) = \mathit{0}
$$

And from the conclusion of [exercise 3.2](exercise/3.2_exercises.ipynb#2.-Let-$T:-V-\to-W$-be-a-linear-map-from-one-vector-space-into-another.-Show-that-$T(\mathit{0})-=-\mathit{0}$.), we have

$$
F(0 v_1 + ... + 0 v_n) = F(\mathit{0}) = \mathit{0}
$$

Since the kernel of $F$ is $\{\mathit{0}\}$, we know $F$ is [injective](#Kernel-and-injectivity). Hence

$$
0 v_1 + ... + 0 v_n = a_1 v_1 + ... + a_n v_n
$$

Thus $a_1 = 0, ..., a_n = 0$. Which leads to a contradiction. Hence $F(v_1), ..., F(v_n)$ are linearly independent in $W$. Q.E.D.

## Image of a linear map as a subspace

Let $F: V \to W$ be a linear map. The image of $F$ is the set of elements $w$ in $W$ such that there exists an element of $v$ of $V$ such that $F(v) = w$.

The image of $F$ is a subspace of $W$.

### Proof

(i) Let $w_1, w_2$ be elements of the image of $F$. Thus there exists $v_1, v_2 \in V$ such that $F(v_1) = w_1$ and $F(v_2) = w_2$. Since $F$ is a linear map, we have $F(v_1 + v_2) = F(v_1) + F(v_2) = w_1 + w_2$. Thus $w_1 + w_2$ is also in the image of $F$.

(ii) Let $w$ be an element of the image of $F$. Thus there exists $v \in V$ such that $F(v) = w$. And let $c$ be a number. Since $F$ is a linear map, we have $F(c v) = c F(v) = c w$. Thus $c w$ is also in the image of $F$.

(iii) From the conclusion of [exercise 3.2](exercise/3.2_exercises.ipynb#2.-Let-$T:-V-\to-W$-be-a-linear-map-from-one-vector-space-into-another.-Show-that-$T(\mathit{0})-=-\mathit{0}$.), we have $F(\mathit{0_{V}}) = F(\mathit{0}_W)$. Hence $\mathit{0_W}$ is in the image of $F$.

Hence the image of $F$ is a subspace of $W$. Q.E.D.

We denote the image of $F$ by $\operatorname{Im} F$.

## Theorem 3.2: Kernel, image, and dimensions

Let $V$ be a vector space. Let $L: V \to W$ be a linear map of $V$ into another space $W$. Let $n$ be the dimension of $V$, $q$ the dimension of the kernel of $L$, and $s$ the dimension of the image of $L$. Then $n = q + s$. In other words,

$$
\dim V = \dim \operatorname{Ker} L + \dim \operatorname{Im} L
$$

### Proof

#### Trival case: $\operatorname{Im} L = \{\mathit{0}\}$

If $\operatorname{Im} L = \{\mathit{0}\}$, since [$\dim \{\mathit{0}\} = 0$](../1_vector_spaces/1_vector_spaces.ipynb#Definition-of-dimensions), we have $\dim \operatorname{Im} L = 0$. In this case, for all $v \in V$ $F(v) = \mathit{0}$. Hence $V = \operatorname{Ker} L$. Thus

$$
\dim V = \dim \operatorname{Ker} L = \dim \operatorname{Ker} L + 0 = \dim \operatorname{Ker} L + \dim \operatorname{Im} L.
$$

#### Non-trival case: $\operatorname{Im} L \neq \{\mathit{0}\}$

If $\operatorname{Im} L \neq \{\mathit{0}\}$, we have $\dim \operatorname{Im} L > 0$. 

Then we let $\{w_1, ..., w_s\}$ be a basis of $\operatorname{Im} L$. And we let $v_1, ..., v_n \in V$ such that $L(v_1) = w_1, ..., L(v_s) = w_s$.

If $\operatorname{Ker} L \neq \{\mathit{0}\}$, let $\{u_1, ..., u_q\}$ be a basis of $\operatorname{Ker} L$. If $\operatorname{Ker} L = \{\mathit{0}\}$, it is understood that all reference to $\{u_l, ..., u_q\}$ is to be omitted in what follows.

The we will contend $\{v_1, ..., v_s, u_1, ..., u_q\}$ is a basis of $V$.

Let $v$ be any element of $V$. Since $\{w_1, ..., w_n\}$ is a basis of $\operatorname{Im} L$, there exists numbers $x_1, ..., x_s$ such that

$$
L(v) = x_1 w_1 + ... + x_s w_s
$$

And since $L(v_1) = w_1, ..., L(v_s) = w_s$, and since $L$ is linear, we have

$$
L(v) = x_1 L(v_1) + ... + x_s L(v_s) = L(x_1 v_1 + ... + x_s v_s)
$$

Thus

$$
L(v - (x_1 v_1 + ... + x_s v_s)) = L(v) - L(x_1 v_1 + ... + x_s v_s) = \mathit{0}
$$

Thus $v - (x_1 v_1 + ... + x_s v_s)$ is in $\operatorname{Ker} L$. And since $\{u_1, .., u_q\}$ is a basis of $\operatorname{Ker} L$, there exists numbers $y_1, ..., y_q$ such that

$$
v - (x_1 v_1 + ... + x_s v_s) = y_1 u_1 + ... + y_q u_q
$$

Thus

$$
v = x_1 v_1 + ... + x_s v_s + y_1 u_1 + ... + y_q u_q
$$

We have now proven any element $v \in V$ is spanned by $v_1, ..., v_s, u_1, ..., u_q$.

To show $v_1, ..., v_s, u_1, ..., u_q$ are linearly independent, we suppose there exists numbers $x_1, ..., x_s, y_1, ..., y_q$ such that

$$
x_1 v_1 + ... + x_s v_s + y_1 u_1 + ... + y_q u_q = \mathit{0}
$$

Thus

$$
L(x_1 v_1 + ... + x_s v_s + y_1 u_1 + ... + y_q u_q) = \mathit{0}
$$

And since $\{u_1, ..., u_q\}$ is a basis of $\operatorname{Ker} L$, we have $L(u_1) = \mathit{0}, ..., L(u_q) =  \mathit{0}$. Hence

$$
\begin{aligned}
L(x_1 v_1 + ... + x_s v_s + y_1 u_1 + ... + y_q u_q) &= x_1 L(v_1) + ... + x_s L(v_s) + y_1 L(u_1) + ... + y_q L(u_1) \\ 
&= x_1 L(v_1) + ... + x_s L(v_s) + \mathit{0} + ... + \mathit{0} \\ 
&= x_1 L(v_1) + ... + x_s L(v_s)
\end{aligned}
$$

Thus

$$
x_1 L(v_1) + ... + x_s L(v_s) = \mathit{0}
$$

And recall that $L(v_1) = w_1, ..., L(v_s) = w_s$ as $\{w_1, ..., w_s\}$ is a basis of $\dim \operatorname{Im} L$, thus $L(v_1), ..., L(v_s)$ are linear independent. Hence $x_1 = 0, ..., x_s = 0$. Hence

$$
y_1 u_1 + ... + y_q u_q = \mathit{0}
$$

And since $\{u_1, ..., u_q\}$ is a basis of $\operatorname{Ker} L$, $u_1, ..., u_1$ are linearly independent, thus $y_1 = 0, ..., y_q = 0$.

Hence $x_1, ..., x_s, y_1, ..., y_q$ all $= 0$. Hence $v_1, ..., v_s, u_1, ..., u_q$ are linearly independent.

Hence $\{v_1, ..., v_s, u_1, ..., u_q\}$ is a basis of $V$. Hence

$$
\dim V = \dim \operatorname{Ker} L + \dim \operatorname{Im} L
$$

Q.E.D.

## Theorem 3.3: Kernel, dimensions, and bijectivity

Let $L: V \to W$ be a linear map. Assume that

$$
\dim V = \dim W
$$

$\operatorname{Ker} L = \{\mathit{0}\}$, or if $\operatorname{Im} L = W$, if and only if $L$ is bijective.

### Proof

#### Forward

If $\operatorname{Ker} L = \{\mathit{0}\}$, by [theorem 3.2](#Theorem-3.2:-Kernel,-image,-and-dimensions) we have $\dim V = \dim \operatorname{Im} L$.

And since $\dim V = \dim W$, we have $\dim \operatorname{Im} L = \dim W$.

And by the [corollary 3.5 of Chapter I](../1_vector_spaces/1_vector_spaces.ipynb#Corollary-3.5.), we have $\operatorname{Im} L = W$. Thus $L$ is surjective. 

And we know $\operatorname{Ker} L = \{0\}$, which is [equivalent](#Kernel-and-injectivity) to $L$ being injective.

Thus $L$ is bijective.

#### Converse

If $L$ is bijective, then $L$ is surjective and injective.

Since $L$ is $injective$, [we have](#Kernel-and-injectivity) $\operatorname{Ker} L = \{0\}$.

Q.E.D.
