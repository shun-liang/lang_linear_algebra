## Theorem 4.1: Composition of two linear maps is also a linear map

Let $U, V, W$ be vector spaces over a field $K$. Let

$$
\begin{aligned}
F: U \to V &&\text{and}&& G: V \to W
\end{aligned}
$$

be linear maps. Then the composition map $G \circ F$ is also a linear map.

### Proof

$G \circ F$ is a map from $U$ to $W$.

#### LM 1

For any $u, v \in U$, since $F, G$ are linear maps, we have

$$
(G \circ F)(u + v) = G(F(u + v)) = G(F(u) + F(v)) = G(F(u)) + G(F(v)) = (G \circ F)(u) + (G \circ F)(v)
$$

#### LM 2

For any $u \in U$, for any $c \in K$, since $F, G$ are linear maps, we have

$$
(G \circ F)(c u) = G(F(c u)) = G(c F(u)) = c G(F(u)) = c ((G \circ F)(u))
$$

Hence $(G \circ F)$ is a linear map. Q.E.D.

## Theorem 4.2. Algebraic rules for linear map compositions

Let $U, V, W$ be vector spaces over a field $K$. Let

$$
F: U \to V
$$

be a linear mapping, and let $G, H$ be two linear mappings of $V$ into $W$.

Then

$$
(G + H) \circ F = G \circ F + H \circ F
$$

If $c$ is a number, then

$$
(c G) \circ F = c (G \circ F)
$$

If $T: U \to V$ is a linear mapping from $U$ into $V$, then

$$
G \circ (F + T) = G \circ F + G \circ T
$$

### Proof

#### Proof: $(G + H) \circ F = G \circ F + H \circ F$
 
For any $u \in U$, we have

$$
((G + H) \circ F)(u) = ((G + H)(F(u)))
$$

By the [definition](3_2_linear_mappings.ipynb#Sum) of sums of linear mappings, we have

$$
((G + H)(F(u))) = G(F(u)) + H(F(u)) = (G \circ F)(u) + (H \circ F)(u)
$$


Thus 

$$
(G + H) \circ F = G \circ F + H \circ F
$$

#### Proof: $(c G) \circ F = c (G \circ F)$

For any $u \in U$, we have

$$
((c G) \circ F)(u) = (c G)(F(u))
$$

By the [definition](3_2_linear_mappings.ipynb#Scalar-multiplication) of scalar products of linear mappings, we have

$$
(c G)(F(u)) = c (G(F(u))) = c ((G \circ F)(u))
$$

Thus

$$
(c G) \circ F) = c (G \circ F)
$$

#### Proof: $G \circ (F + T) = G \circ F + G \circ T$

For any $u \in U$, we have

$$
(G \circ (F + T))(u) = G((F + T)(u)) = G(F(u) + T(u))
$$

And since $G$ is a linear map, we have

$$
G(F(u) + T(u)) = G(F(u)) + G(T(u)) = (G \circ F)(u) + (G \circ T)(u)
$$

Hence

$$
G \circ (F + T) = G \circ F + G \circ T
$$

Q.E.D.

## Compositions of linear mappings with same domains and co-domains

Let $F: U \to U$ and $G: U \to U$ be two linear mappings. Then we may form $F \circ G$ and $G \circ F$.

It's not always true that $F \circ G$ and $G \circ F$ are equal.

### Example: $F \circ G \neq G \circ F$

Let $U = R^3$.

Let $F$ be the linear mapping given by

$$
F(x, y, z) = (x, y, 0)
$$

and $G$ be the linear mapping given by

$$
G(x, y, z) = (x, z, 0)
$$

Then

$$
(F \circ G)(x, y, z) = F(G(x, y, z)) = F(x, z, 0) = (x, z, 0)
$$

and

$$
(G \circ F)(x, y, z) = G(F(x, y, z)) = G(x, y, 0) = (x, 0, 0)
$$

Hence

$$
F \circ G \neq G \circ F
$$

## Operators and their compositions

Let $F: V \to V$ be a linear map of a vector space into itself. One sometimes calls $F$ an **operator**.

Then we can form the composite $F \circ F$, which is again a linear map of $V$ into itself.

We denote the composition of $F$ by itself for $n$ times for any integer $n \geq 1$ as $F^n$, thus

$$
F^n = \underbrace{F \circ F \circ ... \circ F}_{\text{$n$ times}}
$$

And we define

$$
F^0 = I
$$

We have the rules

$$
F^{r + s} = F^r \circ F^s
$$

for integers $r, s \geq 0$.

## Theorem 4.3.

Let $F: U \to V$ be a linear map, and assume that this map has an inverse mapping $G: V \to U$. Then $G$ is a linear map.

### Proof

#### LM 1

For any $v_1, v_2 \in V$, let $u_1 = G(v_1), u_2 = G(v_2)$, and since $G$ is a mapping from $V$ into $U$, then $u_1, u_2$ are elements of $U$.

Since $F$ is a linear map from $U$ into $V$, we have

$$
F(u_1 + u_2) = F(u_1) + F(u_2)
$$

And since $G$ is the inverse of $F$, thus $F \circ G = I$, thus we have

$$
F(u_1) + F(u_2) = F(G(v_1)) + F(G(v_2)) = (F \circ G)(v_1) + (F \circ G)(v_2) = v_1 + v_2
$$

Thus

$$
F(u_1 + u_2) = v_1 + v_2
$$

Passing both sides to $G$, we have

$$
G(F(u_1 + u_2)) = (G \circ F)(u_1 + u_2) = G(v_1 + v_2)
$$

Since $G \circ F = I$, we have

$$
u_1 + u_2 = G(v_1 + v_2)
$$

Thus

$$
G(v_1 + v_2) = G(v_1) + G(v_2)
$$

#### LM 2

For any $v \in V$, let $u = G(v)$. Let $c$ be any number.

Since $F$ is a linear map from $U$ into $V$, we have

$$
F(c u) = c F(u)
$$

And since $G$ is the inverse of $F$, thus $F \circ G = I$, thus we have

$$
c F(u) = c F(G(v)) = c ((F \circ G)(v)) = c v
$$

Thus

$$
F(c u) = c v
$$

Passing both sides to $G$, we have

$$
G(F(c u)) = G(c v)
$$

And since $G$ is the inverse of $F$, thus $G \circ F = I$, thus we have

$$
G(F(c u)) = (G \circ F)(c u) = c v = G(c v)
$$

Hence

$$
G(c v) = c G(v)
$$

Hence $G$ is a linear map. Q.E.D.

## Corollary 4.4.

Let $F: U \to V$ be a linear map whose kernel is $\{\mathit{0}\}$, and which is surjective. Then $F$ has an inverse linear map.

### Proof

#### Proof: the inverse exists

We contend $F$ is injective.

Let $F$ be a non-injective linear map. Then there exists $u_1, u_2 \in U$ such that

$$
u_1 \neq u_2
$$

$$
F(u_1) = F(u_2)
$$

Then we have

$$
F(u_1 - u_2) = F(u_1) - F(u_2) = \mathit{0}
$$

And

$$
u_1 - u_2 \neq \mathit{0}
$$

The kernel of $F$ has more element than $\mathit{0}$, which leads to a contradiction.

Hence $F$ is injective. Since $F$ is also surjective, then $F$ is bijective.

Hence $F$ [has an inverse](3_1_mappings.ipynb#Bijectiveness-and-inverse-existence).

#### Proof: the inverse is linear

And by [theorem 4.3](Theorem-4.3.), as the inverse of $F$ exists, it is also a linear map.

Q.E.D.

### isomorphism

A linear map $F: U \to V$ which has an inverse $G: V \to U$ (we also say invertible) is called an **isomorphism**.

We eventually will write $F G$ instead of $F \circ G$.
