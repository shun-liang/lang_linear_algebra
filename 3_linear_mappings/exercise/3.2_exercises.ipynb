## 1. Determine which of the following mappings $F$ are linear.

### (a) $F: R^3 \to R^2$ defined by $F(x, y, z)= (x, z)$

#### Solution

Verifying **LM 1**.

Let $(x_1, y_1, z_1), (x_2, y_2, z_2) \in R^3$. Then

$$
F(x_1 + x_2, y_1 + y_2, z_1 + z_2) = (x_1 + x_2, z_1 + z_2) = (x_1, z_1) + (x_2, z_2) = F(x_1, y_1, z_1) + F(x_2, y_2, z_2)
$$

Verifying **LM 2**.

Let $(x, y, z)\in R^3$. Let $c \in R$. Then

$$
F(c x, c y, c z)= (c x, c z) = c(x, z) = c F(x, y, z)
$$

Hence $F$ is a linear mapping.

### (b) $F: R^4 \to R^4$ defined by $F(X)= - X$

#### Solution

Verifying **LM 1**.

Let $(x_1, y_1, z_1, t_1), (x_2, y_2, z_2, t_2) \in R^4$. Then

$$
\begin{equation}
\begin{aligned}
F((x_1, y_1, z_1, t_1) + (x_2, y_2, z_2, t_2)) &= F((x_1 + x_2, y_1 + y_2, z_1 + z_2, t_1 + t_2)) \\
&= - (x_1 + x_2, y_1 + y_2, z_1 + z_2, t_1 + t_2) \\
&= (- (x_1, y_2, z_1, t_1)) + (- (x_2, y_2, z_2, t_2)) \\
&= F((x_1, y_1, z_1, t_1)) + F((x_2, y_2, z_2, t_2)) \\
\end{aligned}
\end{equation}
$$

Verifying **LM 2**.

Let $(x, y, z, t) \in R^4$. Let $c \in R$. Then

$$
\begin{equation}
\begin{aligned}
F(c (x, y, z, t)) = F((c x, c y , c z, c t)) = - (c x, c y , c z, c t) = c (- (x, y, z, t)) = c F((x, y, z, t))
\end{aligned}
\end{equation}
$$

Hence $F$ is a linear mapping.

### (c) $F: R^3 \to R^3$ defined by $F(X) = X + (0, -1, 0)$

#### Solution

Verifying **LM 1**

Let $(x_1, y_1, z_1), (x_2, y_2, z_2) \in R^3$. Then

$$
\begin{equation}
\begin{aligned}
F((x_1, y_1, z_1) + (x_2, y_2, z_2)) &= F((x_1 + x_2, y_1 + y_2, z_1 + z_2)) \\
&= (x_1 + x_2, y_1 + y_2, z_1 + z_2) + (0, -1 ,0) \\
&\neq ((x_1, y_1, z_1) + (0, -1, 0)) + ((x_2, y_2, z_2) + (0, -1, 0)) \\
&= F((x_1, y_1, z_1)) + F((x_2, y_2, z_2))
\end{aligned}
\end{equation}
$$

Hence **LM 1** does not hold for $F$.

Hence $F$ is **not** a linear mapping.

### (d) $F: R^2 \to R^2$ defined by $F(x, y) = (2x + y, y)$

#### Solution

Verifying **LM 1**.

Let $(x_1, y_1), (x_2, y_2) \in R^2$. Then

$$
F(x_1, y_1) + (x_2, y_2) = F(x_1 + x_2, y_1 + y_2) = (2 (x_1 + x_2), y_1 + y_2) = (2 x_1 + 2 x_2, y_1 + y_2) = (2 x_1, y_1) + (2 x_2, y_2) = F(x_1, y_1) + F(x_2, y_2)
$$

Verifying **LM 2**.

Let $(x, y) \in R^2$. Let $c \in R$. Then

$$
F(c x , c y) = (2 (c x + c y), c y) = (c (2 (x + y)), c y) = c (2 (x + y), y) = c F(x, y)
$$

Hence $F$ is a linear mapping.

### (e) $F:R^2 \to R^2$ defined by $F(x,y) = (2 x, y - x)$

#### Solution

Verifying **LM 1**.

Let $(x_1, y_1), (x_2, y_2) \in R^2$. Then

$$
\begin{equation}
\begin{aligned}
F(x_1, y_1) + (x_2, y_2) &= F(x_1 + x_2, y_1 + y_2) \\
&= (2 (x_1 + x_2), (y_1 + y_2) - (x_1 + x_2)) \\
&= (2 x_1 + 2 x_2, (y_1 - x_1) + (y_2 - x_2)) \\
&= (2 x_1, y_1 - x_1) + (2 x_2, y_2 - x_2) \\
&= F(x_1, y_1) + F(x_2, y_2)
\end{aligned}
\end{equation}
$$

Verifying **LM 2**.

Let $(x, y) \in R^2$. Let $c \in R$. Then

$$
F(c x, c y) = (2 c x, c y - c x) = c (2 x , y - x) = c F(x, y)
$$

Hence $F$ is a linear mapping.

### (g) $F:R^2 \to R$ defined by $F(x, y)=x y$

#### Solution

Verifying **LM 1**.

Let $(x_1, y_1), (x_2, y_2) \in R^2$. Then

$$
\begin{equation}
\begin{aligned}
F(x_1 + x_2, y_1 + y_2) &= (x_1 + x_2) (y_1 + y_2) \\
&= x_1 y_1 + x_2 y_1 + x_1 y_2 + x_2 y_2 \\
&\neq x_1 y_1 + x_2 y_2 \\
&= F(x_1, y_1) + F(x_2, y_2)
\end{aligned}
\end{equation}
$$

Hence **LM 1** does not hold.

Hence $F$ is **not** a linear mapping.

### (h) Let $U$ be an open subset of $R^3$, and let $V$ be the vector space of differentiable functions on $U$. Let $V^\prime$ be the vector space of vector fields on $U$. Then $\text{grad}: V \to V^\prime$ is a mapping. Is it linear? (For this part (h) we assume you know some calculus.)

#### Solution

Let $x, y, z$ denotes the coordinates of $R^3$.

For any $F \in V$, [the definition](https://en.wikipedia.org/wiki/Gradient#Cartesian_coordinates) of $\text{grad}(F)$ is

$$
\text{grad}(F) = (\frac{\partial F}{x}, \frac{\partial F}{y}, \frac{\partial F}{z})
$$


Verifying **LM 1**.

Let $F, G \in V$. Then we have

$$
\begin{equation}
\begin{aligned}
\text{grad}(F + G) &= (\frac{\partial (F + G)}{x}, \frac{\partial (F + G)}{y}, \frac{\partial (F + G)}{z}) \\
&= (\frac{\partial F}{x} + \frac{\partial G}{x}, \frac{\partial F}{y} + \frac{\partial G}{y}, \frac{\partial F}{z} + \frac{\partial G}{z}) \\
&= (\frac{\partial F}{x}, \frac{\partial F}{y}, \frac{\partial F}{z}) + (\frac{\partial G}{x}, \frac{\partial G}{y}, \frac{\partial G}{z}) \\
&= \text{grad}(F) + \text{grad}(G)
\end{aligned}
\end{equation}
$$

Verifying **LM 2**.

Let $F \in V$. Let $c \in R$. Then we have


$$
\begin{equation}
\begin{aligned}
\text{grad}(c F) &= (\frac{\partial (c F)}{x}, \frac{\partial (c F)}{y}, \frac{\partial (c F)}{z}) \\
&= (c \frac{\partial F}{x}, c \frac{\partial F}{y}, c \frac{\partial F}{z}) \\
&= c (\frac{\partial F}{x}, \frac{\partial F}{y}, \frac{\partial F}{z}) \\
&= c \cdot \text{grad}(F)
\end{aligned}
\end{equation}
$$


Hence $\text{grad}$ is a linear mapping. Q.E.D.

Thus $t(h) = w$. Thus $S \subseteq (v_0 + U)$.

### Generalization


Let $V$ be a vector space, and for some $n \in \mathbb{N}$, let $f_1: V \to R, f_2: V \to R, ..., f_n: V \to R$ be $n$ linear mappings. Let $F: V \to R^n$ be the mapping defined by $F(v) = (f_1(v), f_2(v), ..., f_n(v))$. Show that $F$ is linear.

#### Proof

Verify **LM 1**.

Let $u, w$ be any two elements in $V$. Since $f_i$ is a linear map for $i = 1, ..., n$, then

$$
\begin{equation}
\begin{aligned}
F(u + w) &= (f_1(u + w), f_2(u + w), ..., f_n(u + w)) \\
&= (f_1(u) + f_1(w), f_2(u) + f_2(w), ..., f_n(u) + f_n(w)) \\
&= (f_1(u), f_2(u), ..., f_n(u)) + (f_1(w), f_2(w), ..., f_n(w)) \\
&= F(u) + F(w)
\end{aligned}
\end{equation}
$$

Verify **LM 2**.

Let $v$ be any element in $V$, and let $c$ be any number in $R$. Since $f_i$ is a linear map for $i = 1, ..., n$, then

$$
\begin{equation}
\begin{aligned}
F(c v) &= (f_1(c v), f_2(c v), ..., f_n(c v)) \\
&= (c f_1(v), c f_2(v), ..., c f_n(v)) \\
&= c (f_1(v), f_2(v), ..., f_n(v))
\end{aligned}
\end{equation}
$$

## 9.

Let $V$ be a vector space over $R$, and let $v, w \in V$. The **line passing through $v$ and parallel to $w$** is defined to be the set of all elements $v + t w$ with $t \in R$. The line segment between $v$ and $v + w$ is defined to be the set of all elements


$$
\begin{equation}
\begin{aligned}
v + t w && \text{with} && 0 \leq t \leq 1
\end{aligned}
\end{equation}
$$

### (i)

Let $L: V \to U$ be a linear map. Show that the image under $L$ of a line segment in $V$ is a line segment in $U$. Between what points?

#### Solution

Let 

$$
S = \{v + t w | v, w \in V, 0 \leq t \leq 1 \}
$$

Let $u \in S$, then there exists some $t \in R$ where $0 \leq t \leq 1$ such that

$$
u = v + t w
$$

Thus

$$
f(u) = L(v + t w) = L(v) + L(t w)= L(v) + t L(w)
$$

Let $S^\prime$ be subset of $U$ such that **line passing through $f(v)$ and parallel to $L(w)$**. Then we have $L(S) \subseteq S^\prime$.

For any $s \in S^\prime$, there exists some some $t \in R$ $0 \leq t \leq 1$ such that $s^\prime = L(v) + t L(w)$

Hence

$$
s = L(v) + t L(w) = L(v) + L(t w) = L(v + t w)
$$

Hence $S^\prime \subseteq f(S)$.

Hence $f(S) = S^\prime$. Hence the image under $L$ of a line segment in $V$ between $v, w$ is a line segment in $U$ that pass through $L(v)$ and is parallel to $L(w)$.

### (ii) Show that the image of a line under $L$ is either a line or a point.

### Proof

From the solution at (i), for $v, w \in V$, we can see the image of the line passing through $v$ and is parallel to $w$ is

$$
L(v) + t L(w)
$$

for $t \in R$.

When $L(w) \neq 0$, the image of the line under $L$ is a line passing through $L(v)$ and is parallel to $L(w)$.

When $L(w) = 0$, the image of the line under $L$ is the $L(v)$.

Q.E.D.

## Definition: parallelogram


Let $V$ be a vector space, and let $v_1, v_2$ be two elements of V which are linearly independent. The set of elements of $V$ which can be written in the form 

$$
t_1 v_1 + t_2 v_2
$$

with numbers $t_1, t_2$ satisfying

$$
\begin{aligned}
0 \leq t_1 \leq 1 && \text{and} && 0 \leq t_2 \leq 1
\end{aligned}
$$

is called the **parallelogram** spanned by $v_1, v_2$.

## 10. Let $V$ and $W$ be vector spaces, and let $F: V \to W$ be a linear map. Let $v_1, v_2$ be linearly independent elements of $V$, and assume that $F(v_1), F(v_2)$ are linearly independent. Show that the image under $F$ of the parallelogram spanned by $v_1$ and $v_2$ is the parallelogram spanned by $F(v_1), F(v_2)$.

### Proof

Let $P$ be the parallelogram spanned by $v_1$ and $v_2$. Thus for any $u \in P$, we have $v = t_1 v_1 + t_2 v_2$ for some $0 \leq t_1 \leq 1$ and $0 \leq t_2 \leq 1$. As $F$ is a linear map, we have

$$
F(u)= F(t_1 v_1 + t_2 v_2) = F(t_1 v_1) + F(t_2 v_2) = t_1 F(v_1) + t_2 F(v_2)
$$

Let $P^\prime$ be the parallelogram spanned by $F(v_1)$ and $F(v_1)$. Then from above we have $F(P) \subseteq P^\prime$.

For any $w \in P^\prime$, we have $w = t_1 F(v_1) + t_2 F(v_2)$ for some $0 \leq t_1 \leq 1$ and $0 \leq t_2 \leq 1$. Since $F$ is a linear map, we have

$$
w = t_1 F(v_1) + t_2 F(v_2) = F(t_1 v_1) + F(t_2 v_2) = F(t_1 v_1 + t_2 v_2)
$$

Hence $P^\prime \subseteq F(P)$.

Hence $F(P) = S^\prime$. Thus the image under $F$ of the parallelogram spanned by $v_1$ and $v_2$ is the parallelogram spanned by $F(v_1), F(v_2)$. Q.E.D.

## 15. Let $V, W$ be two vector spaces, and $F: V \to W$ a linear map. Let $w_1, ... ,w_n$ be elements of $W$ which are linearly independent, and let $v_1, ..., v_n$ be elements of $V$ such that $F(v_i) = w_i$ for $i = 1,...,n$.  Show that $v_1, ...,v_n$ are linearly independent.

### Proof

Assume $v_1, ..., v_n$ are linearly dependent. Thus there exists $a_1, ..., a_n$ which are not all $= 0$ such that

$$
a_1 v_1 + ... + a_n v_n = \mathit{0}
$$

Thus

$$
F(a_1 v_1 + ... + a_n v_n) = \mathit{0}
$$

Since $F$ is a linear map, we have

$$
F(a_1 v_1) + ... + F(a_n v_n) = \mathit{0} \\
a_1 F(v_1) + ... + a_n F(v_n) = \mathit{0} \\
a_1 w_1 + ... + a_n w_n = \mathit{0}
$$

Since $w_1, ..., w_n$ are linearly independent, we have $a_1, ..., a_n$ all $= 0$, which leads to a contradiction. Hence $v_1, ...,v_n$ are linearly independent. Q.E.D.

## 16. Let $V$ be a vector space and $F:V \to R$ a linear map. Let $W$ be the subset of $V$ consisting of all elements $v$ such that $F(v) = 0$. Assume that $W \neq V$, and let $v_0$ be an element of $V$ which does not lie in $W$. Show that every element of $V$ can be written as a sum $w + c v_0$, with some $w$ in $W$and some number $c$.

### Proof

Let $v$ be any element in $V$.

Since $F$ maps to $R$, and since there exists some $v_0$ such that $F(v_0) \neq 0$, we let $c = \frac{F(v)}{F(v_0)}$.

Let $w = v - c v_0$. Since $F$ is a linear map, we have

$$
F(w) = F(v - c v_0) = F(v) - c F(v_0) = F(v) - \frac{F(v)}{F(v_0)} F(v_0) = F(v) - F(v) = 0
$$

Thus $F(w) = 0$. Hence $w \in W$.

Hence every element of $V$ can be written as a sum $w + c v_0$, with some $w$ in $W$and some number $c$. Q.E.D.

## 17. In Exercise 16, show that $W$ is a subspace of $V$. Let $\{v_1, ... ,v_n\}$ be a basis of $W$. Show that $\{v_0, v_1, ... ,v_n\}$ is a basis of $V$.

### Proof: $W$ is a subspace of $V$

(i) Let $w_1, w_2 \in W$. Since $F$ is a linear map, we have

$$
F(w_1 + w_2) = F(w_1) + F(w_2) = 0 + 0 = 0
$$

Hence $w_1 + w_2 \in W$.

(ii) Let $w \in W$ and $c \in R$. Since $F$ is a linear map, we have

$$
F(c w) = c F(w) = c \cdot 0 = 0
$$

(iii) From the conclusion of [exercise 2](#2.-Let-$T:-V-\to-W$-be-a-linear-map-from-one-vector-space-into-another.-Show-that-$T(\mathit{0})-=-\mathit{0}$.), we have $F(\mathit{0}) = 0$. Hence $\mathit{0} \in W$.

Hence $W$ is a subspace of $V$.

### Proof: $\{v_0, v_1, ... ,v_n\}$ is a basis of $V$.

Since $\{v_1, ... ,v_n\}$ be a basis of $W$, any $w \in W$ can be written as

$$
a_1 v_1, ..., a_n v_n
$$

for some unique $a_1, ..., a_n \in R$.

From the conclusion of [exercise 16](#16.-Let-$V$-be-a-vector-space-and-$F:V-\to-R$-a-linear-map.-Let-$W$-be-the-subset-of-$V$-consisting-of-all-elements-$v$-such-that-$F(v)-=-0$.-Assume-that-$W-\neq-V$,-and-let-$v_0$-be-an-element-of-$V$-which-does-not-lie-in-$W$.-Show-that-every-element-of-$V$-can-be-written-as-a-sum-$w-+-c-v_0$,-with-some-$w$-in-$W$and-some-number-$c$.), we know every element in $V$ can be written as $w + c v_0$ for some $w \in W$ and some number $c$.

Thus for every $v \in V$, we have a unique list $c, a_1, ..., a_n \in R$ such that

$$
v = c v_0 a_1 v_1 + ... + a_n v_n
$$

Hence $\{v_0, v_1, ..., v_n\}$ is a basis of $V$.
