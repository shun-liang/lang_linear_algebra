## 1. Let $L: R^2 \to R^2$ be a linear map such that $L \neq \mathit{0}$ but $L^2 = L \circ L = \mathit{0}$. Show that there exists a basis $\{A, B\}$ of $R^2$ such that

$$
\begin{aligned}
L(A) = B && \text{and} && L(B) = \mathit{0}
\end{aligned}
$$

### Proof

Since $L \neq \mathit{0}$, we let $A$ be any vector in $R^2$ such that $A \neq \mathit{0}$ and $L(A) \neq \mathit{0}$. Let $B = L(A)$. We also have $B \neq 0$.

Then we have

$$
L(B) = L(L(A)) = L^2(A) = \mathit{0}(A) = \mathit{0}
$$

Now we contend $\{A, B\}$ is a basis of $R^2$.

Let $\alpha, \beta$ be numbers in $R$ such that

$$
\alpha A + \beta B = \mathit{0}
$$

Then we have

$$
L(\alpha A + \beta B) = L(\mathit{0})
$$

And since [the linear mapping of zero vector is also a zero vector](../3_2_linear_mappings.ipynb#Linear-mapping-at-zero-vector-$\mathit{0}$), we have

$$
L(\alpha A + \beta B) = \mathit{0}
$$

Hence

$$
\alpha L(A) + \beta L(B) = \alpha L(A) + \beta L^2(A) = \alpha L(A) = \mathit{0}
$$

Thus

$$
\alpha L(A) = \alpha B = \mathit{0}
$$

And since $B \neq \mathit{0}$, we have

$$
\alpha = 0
$$

Now we know the only $\alpha, \beta$ for

$$
\alpha A + \beta B = \mathit{0}
$$

to hold is $\alpha = 0, \beta = 0$.

Hence $A, B$ are linear independent.

And since $\dim R^2 = 2$, by [theorem 3.3 in chapter 1](../../1_vector_spaces/1_vector_spaces.ipynb#Theorem-3.3.) we know $\{A, B\}$ is a basis of $R^2$.

Q.E.D.

## 2.  Let $\dim V > \dim W$. Let $L: V \to W$ be a linear map. Show that the kernel of $L$ is not $\{\mathit{0}\}$.

### Proof

From [theorem 3.3](../3_3_kernel_and_image_of_a_linear_map.ipynb#Theorem-3.2:-Kernel,-image,-and-dimensions), we know

$$
\dim V = \dim \operatorname{Ker} L + \dim \operatorname{Im} L
$$

And since $\operatorname{Im} L$ [is a subspace of](../3_3_kernel_and_image_of_a_linear_map.ipynb#Image-of-a-linear-map-as-a-subspace) $W$, by [theorem 3.1 in chapter 1](../../1_vector_spaces.ipynb#Theorem-3.7.) we have

$$
\dim \operatorname{Im} L \leq \dim W
$$

And since $\dim V > \dim W$, we know $\dim V - \dim W > 0$. Hence

$$
\dim \operatorname{Ker} L = \dim V - \dim \operatorname{Im} L \geq \dim V - \dim W > 0
$$

Hence

$$
\dim \operatorname{Ker} L > 0
$$

Hence

$$
\operatorname{Ker} L \neq \{0\}
$$

Q.E.D.

## 5. Let $F, G$ be invertible linear maps of a vector space $V$ onto itself. Show that

$$
(F \circ G)^{-1} = G^{-1} \circ F^{-1}
$$

### Proof

#### Lemma 5.1

Linear mapping composition is associative. A.k.a

Let $F, G, H$ be invertible linear maps of a vector space $V$ onto itself. We have

$$
(F \circ G) \circ H = F \circ (G \circ H)
$$

##### Proof: Lemma 5.1

Let $v$ be any element of $V$. We have

$$
((F \circ G) \circ H)(v) = (F \circ G)(H(v)) = F(G(H(v))) = F((G \circ H)(v)) = (F \circ (G \circ H))(v)
$$

Hence

$$
(F \circ G) \circ H = F \circ (G \circ H)
$$

#### Proof: $(F \circ G)^{-1} = G^{-1} \circ F^{-1}$

From [Lemma 5.1](#Proof:-Lemma-5.1), we have

$$
\begin{aligned}
(F \circ G) \circ (G^{-1} \circ F^{-1}) &= ((F \circ G) \circ G^{-1}) \circ F^{-1} \\
&= (F \circ (G \circ G^{-1})) \circ F^{-1} \\
&= (F \circ I) \circ F^{-1} \\
&= F \circ F^{-1} \\
&= I
\end{aligned}
$$

and

$$
\begin{aligned}
(G^{-1} \circ F^{-1}) \circ (F \circ G) &= ((G^{-1} \circ F^{-1}) \circ F) \circ G \\
&= (G^{-1} \circ (F^{-1} \circ F)) \circ G \\
&= (G^{-1} \circ I) \circ G \\
&= G^{-1} \circ G \\
&= I
\end{aligned}
$$

Hence $G^{-1} \circ F^{-1}$ is the inverse of $F \circ G$. Q.E.D.

## 6. 

Let $L: R^2 \circ R^2$ be the linear map defined by

$$
L(x, y) = (x + y, x - y)
$$

Show that $L$ is invertible.

### Proof

#### Proof by concrete example

If such an inverse exist, the composition of $L$ and the inverse should return $(x, y)$. And in order to find the relationship between $(x, y)$ and $(x + y, x - y)$ we observe that

$$
\begin{cases}
x &= \frac{(x + y) + (x - y)}{2} \\
y &= \frac{(x + y) - (x - y)}{2}
\end{cases}
$$

Let $L^\prime: R^2 \circ R^2$ be map defined by

$$
L^\prime(x, y) = (\frac{x + y}{2}, \frac{x - y}{2})
$$

It's obvious that $L^{\prime}$ is a linear map.

Then we have

$$
(L \circ L^\prime)(x, y) = L(L^\prime(x, y)) = L(\frac{x + y}{2}, \frac{x - y}{2}) = (\frac{x + y}{2} + \frac{x - y}{2}, \frac{x + y}{2} -\frac{x - y}{2}) = (x, y)
$$

$$
(L^\prime \circ L)(x, y) = L^\prime(L(x, y)) = L^\prime(x + y, x - y) = (\frac{(x + y) + (x - y)}{2}, \frac{(x + y) - (x - y)}{2}) = (x, y)
$$

Hence

$$
L \circ L^\prime = L^\prime \circ L = I
$$

Hence $L$ is invertible. Q.E.D.

#### Proof by abstract theorems

I guess we don't always have the fortune to easily find a good candidate of the inverse function. So let's try to follow a more abstract approach here.

##### Proof: $\operatorname{Ker} L = \{\mathit{0}\}$

We contend $\operatorname{Ker} L = \{\mathit{0}\}$.

Let $(x^\prime, y^\prime) \in R^2$ such that 

$$
L(x^\prime, y^\prime) = (x^\prime + y^\prime, x^\prime - y^\prime) = (0, 0)
$$

Then we have

$$
\begin{cases}
x^\prime + y^\prime &= 0 \\
x^\prime - y^\prime &= 0 \\
\end{cases}
$$

Then we have

$$
(x^\prime, y^\prime) = (0, 0)
$$

Hence $\operatorname{Ker} L= \{\mathit{0}\}$.

##### Proof: $\dim \operatorname{Im} L = \dim R^2$

Since $\operatorname{Ker} L = \{\mathit{0}\}$, we know $\dim \operatorname{Ker} L = 0$. By [theorem 3.3](../3_3_kernel_and_image_of_a_linear_map.ipynb#Theorem-3.2:-Kernel,-image,-and-dimensions), we have


Hence
$$
\dim R^2 - \dim \operatorname{Im} L = \dim \operatorname{Ker} L = 0
$$

Thus

$$
\dim R^2 = \dim \operatorname{Im} L
$$

Since $L$ is a linear map from $R^2$ into $R^2$, and $\dim R^2 = \dim \operatorname{Im} L$, and $\operatorname{Ker} L= \{\mathit{0}\}$, by the conclusion of [exercise 4](#4.-Let-$\dim-V-=-\dim-W$.-Let-$L:V-\to-W$-be-a-linear-map-whose-kernel-is-$\{\mathit{0}\}$.-Show-that-$L$-has-an-inverse-linear-map.), we know $L$ is invertible. \operatorname{Ker} L = \{\mathit{0}\}

### (a) Let $L:V \to V$ be a linear mapping such that $L^2 = \mathit{0}$ .Show that $I - L$ is invertible. ($I$ is the identity mapping on $V$.)

#### Proof

By theorem [3.4](../3_4_composition_and_inverse_of_linear_mappings.ipynb#Theorem-4.2.-Algebraic-rules-for-linear-map-compositions), we have

$$
\begin{aligned}
(I - L) \circ (I + L) &= (I - L) \circ I + (I - L) \circ L \\
&= (I - L) + (I \circ L - L^2) \\
&= (I - L) + (L - \mathit{0}) \\
&= (I - L) + L \\
&= I
\end{aligned}
$$

and

$$
\begin{aligned}
(I + L) \circ (I - L) &= (I + L) \circ I - (I + L) \circ L \\
&= (I + L) - (I \circ L + L^2) \\
&= (I + L) - (L + \mathit{0}) \\
&= I
\end{aligned}
$$

Hence $I - L$ is invertible as its inverse is $I + L$.

Q.E.D.

## 10.

Let $V$ be a vector space. Let $P:V \to V$ be a linear map such that $P^2 = P$. Show that

$$
\begin{aligned}
V = \operatorname{Ker} P + \operatorname{Im} P && \text{and} && \operatorname{Ker} P \cap \operatorname{Im} P = \{\mathit{0}\}
\end{aligned}
$$

in other words, $V$ is the direct sum of $\operatorname{Ker} P$ and $\operatorname{Im} P$. \[Hint: To show $V$ is the sum, write an element of $V$ in the form $v = v - P v + P v$.\]

**Note**: $P$ is called a [projection](https://en.wikipedia.org/wiki/Projection_(linear_algebra)) if $P^2 = P$.

### Proof

For any $v \in V$, since $P$ is a linear map from $V$ into $V$, we can write $v$ as

$$
v = v - P v + P v
$$

#### Proof: $\{v^\prime \in V | [\exists v \in V] v^\prime = v - P v\} = \operatorname{Ker} P$

Let $V^\prime = \{v^\prime \in V | [\exists v \in V] v^\prime = v - P v\}$.

We first contend $V^\prime = \operatorname{Ker} P$ by showing $V^\prime \subseteq \operatorname{Ker} P$ and $\operatorname{Ker} P \subseteq V^\prime$.

For any $v^\prime \in V^\prime$, there exists some $v \in V$ such that

$$
P v^\prime = P(v - P v) = P v - P^2 v = P v - P v = \mathit{0}
$$

Hence $V^\prime \subseteq \operatorname{Ker} P$.

Let $v^\prime \in \operatorname{Ker} P$. Then we have

$$
P v^\prime = \mathit{0}
$$

Thus

$$
v^\prime = v^\prime - P v^\prime
$$

And since $\operatorname{Ker} P \subseteq V$, thus $v^\prime \in V$.

Thus $\operatorname{Ker} P \subseteq V^\prime$.

Hence $V^\prime = \operatorname{Ker} P$.

#### Proof: $\{v^{\prime\prime} \in V | [\exists v \in V] v^{\prime\prime} = P v\} = \operatorname{Im} P$

This is obvious.

Thus for any element $v \in V$ can be written as a sum of $v^\prime + v^{\prime\prime}$ for $v^\prime \in \operatorname{Ker} P$ and $v^{\prime\prime} \in \operatorname{Im} P$. 

Thus $V$ the direct sum of $\operatorname{Ker} P$ and $\operatorname{Im} P$ Q.E.D.
