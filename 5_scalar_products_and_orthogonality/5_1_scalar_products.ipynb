## Definition: Scalar Products


Let $V$ be a vector space over a field $K$. A **scalar product** on $V$ is an association which to any pair of elements $v, w$ of $V$ associates a scalar, denoted by $\langle v, w \rangle$, or also $v \cdot w$, satisfying the following properties:

- **SP 1**. We have $\langle v, w \rangle = \langle w, v \rangle$ for all $v, w \in V$.
- **SP 2**. If $u, v, w$ are elements of $V$, then
$$
\langle u, v + w \rangle = \langle u, v \rangle + \langle u, w \rangle
$$
- **SP 3**. If $x \in K$, then
$$
\begin{aligned}
\langle x u, v \rangle = x \langle u, v \rangle && \text{and} && \langle u, x v \rangle = x \langle u, v \rangle 
\end{aligned}
$$

### Non-degenerate scalar products

A scalar is called **non-degenerate** if in addition it also satisfies the condition:

If $v$ is an element of $V$, and $\langle v, w\rangle = 0$ for all $w \in V$, then $v = \mathit{}0$.

### Lemma 1.0

Let $V$ be a vector space over a field $K$ with a scalar product. Then for any $v \in V$, we have

$$
\langle v, \mathit{0} \rangle = 0
$$

#### Proof

Let $w$ be some vectors in $V$. Then we have

$$
\langle v, \mathit{0} \rangle = \langle v, w - w \rangle
$$

By **SP 2**, we have

$$
\langle v, w - w \rangle = \langle v, w \rangle + \langle v, - w \rangle
$$

And by **SP 3**, we have

$$
\langle v, w \rangle + \langle v, - w \rangle = \langle v, w \rangle - \langle v, w \rangle = 0
$$

Hence

$$
\langle v, \mathit{0} \rangle = 0
$$

Q.E.D.

### Example 1: Dot products

Let $V = K^n$. Then the map

$$
(X, Y) \mapsto X \cdot Y
$$

which to elements $X, Y \in K^n$ associates their [dot product](../1_vector_spaces/1_vector_spaces.ipynb#Dot-Products-/-Scalar-Products) as we defined it previously, is a scalar product in the present sense.

### Example 2: Integrals

Let $V$ be the space of continuous real-valued functions on the interval $[0, 1]$. If $f, g \in V$, we define

$$
\langle f, g \rangle = \int_0^1 f(t) g(t) dt
$$

Simple properties of the integral show that this is a scalar product.

## Definition: Orthogonality

Let $V$ be a vector space with a scalar product. We define $v, w$ of $V$ to be **orthogonal** or **perpendicular**, and write $v \perp w$, if $\langle v, w \rangle = 0$.

### Definition: Orthogonal space

Let $S$ be a subset of $V$. We denote $S^\perp$ to be the set of all elements $w \in V$ such that, for all $v \in S$, we have $w, v$ being orthogonal.

Then by **SP 2**, for any $w_1, w_2 \in S^\perp$, we have we have $w_1 + w_2 \in S^\perp$. Thus (i) is satisfied.

And by **SP 3**, for any $w \in S^\perp$, and any number $c$, we have $c w \in S^\perp$. Thus (ii) is satisfied.

And by **SP 2** again, let $w$ be some element in $S^\perp$. Then we have $w + (- w) = \mathit{0} \in S^\perp$. Thus (iii) is satisfied.

Since (i), (ii), and (iii) are all satisfied, then [we know](../1_vector_spaces/1_vector_spaces.ipynb#Subspaces) $S^\perp$ is a subspace of $V$. Then we can call $S^\perp$ the **orthogonal space** of $S$.

For any $w \in S^\perp$, we can write $w \perp S$, reads $w$ is orthogonal to $S$.

Let $U$ be the subspace of $V$ generated by the elements of $S$. If $w \perp S$, and if $v_1, v_2 \in S$, then by **SP 2**

$$
\langle w, v_1 + v_2 \rangle = \langle w, v_1 \rangle + \langle w, v_2 \rangle = 0
$$

And let $c$ be any number. By **SP 3**

$$
\langle w, c v_1 \rangle = c \langle w, v_1 \rangle
$$

Hence $w$ is orthogonal to linear combinations of the elements of $S$. Hence $w \perp U$.

## Example 3.

Let $(a_{ij})$ be an $m \times n$ matrix in $K$, and let $A_1, ... , A_m$ be its row vectors. Let $X = \begin{pmatrix}x_1 \\ \vdots \\ x_n \end{pmatrix}$. The system of homogeneous linear equations

$$
(**) \begin{array}{c}
a_{1 1} x_1 + \cdots + a_{1 n} x_n = 0 \\
\vdots \\
a_{m 1} x_1 + \cdots + a_{m n} x_n = 0
\end{array}
$$

can also be written in an abbreviated form

$$
\begin{array}{c}
A_1 \cdot X &= 0 \\
\vdots \\
A_m \cdot X &= 0
\end{array}
$$

The set of solutions $X$ of this homogeneous system is a vector space over $K$.

In fact, let $W$ be the space generated by $A_1, ..., A_m$. Let $U$ be the space consisting of all vectors in $K^n$ orthogonal to $A_1, ..., A_m$. Then $U$ is precisely the vector space of solutions to (\*\*). The vectors $A_1, ..., A_m$ may not be linearly independent. We note that $\dim W \leq m$, and we call

$$
\dim U = \dim W^\perp
$$

the **dimension of the space of solutions of the system of linear equations**. We shall discuss this dimension at greater length later.

### Orthogonal Basis

Let $V$ be a vector space over the field $K$, with a scalar product.

Let $\{v_1, ..., v_n\}$ be a basis of V. We shall say that it is an **orthogonal basis** if $\langle v_i, v_j \rangle = 0$ for all $i \neq j$.

We shall show later that if $V$ is a finite dimensional vector space, with a scalar product, then there always exists an orthogonal basis. However, we shall first discuss important special cases over the real and complex numbers.

## The real positive definite case

Let $V$ be a vector space over $R$, with a scalar product. We shall call this scalar product **positive definite** if

- $\langle v, v \rangle \geq 0$ for all $v \in V$, 
- and $\langle v, v \rangle > 0$ if $v \neq \mathit{0}$. 

The ordinary dot product of vectors in $R^n$ is positive definite, and so is the scalar product of Example 2 above.

Let $V$ be a vector space over $R$, with a positive definite scalar product denoted by $\langle, \rangle$. Let $W$ be a subspace. Then $W$ has a scalar product defined by the same rule defining the scalar product in $V$. In other words, if $w, w^\prime$ are elements of $W$, we may form their product $\langle w, w \rangle$. This scalar product on $W$ is obviously positive definite.

## Definitions: Norm and distance

In positive definite case, we have

$$
\langle v, v \rangle \geq 0
$$

Then we can define the **norm** of an vector $v \in V$ by

$$
\Vert v \Vert = \sqrt{\langle v, v \rangle}
$$

If $c$ is any number, then we get

$$
\Vert c v \Vert = \sqrt{\langle c v, c v \rangle} = \sqrt{c \langle v, c v \rangle} = \sqrt{c^2 \langle v, v \rangle} = c \sqrt{\langle v, v \rangle} = c \Vert v \Vert
$$

The **distance** between two elements $v, w$ of $V$ is defined to be

$$
\operatorname{dist}(v, w) = \Vert v - w \Vert
$$

This definition stems from the Pythagoras theorem. For example, suppose $V = R^3$ with the ordinary dot product as the scalar product. If $X = \begin{pmatrix} x \\ y \\ z \end{pmatrix} \in V$, then

$$
\Vert X \Vert = \sqrt{x^2 + y^2 + z^2}
$$


This coincides precisely with our notion of distance from the origin $\mathit{0}$ to the point $A$ by making use of Pythagoras' theorem.


We can also justify our definition of orthogonality. Again the intuition of plane geometry and the following figure tell us that $v$ is orthogonal to $w$ if and only if

$$
\Vert v - w \Vert = \Vert v + w \Vert
$$

![figure 1](figure_1.png)

Meanwhile, by algebra:

$$
\begin{aligned}
\Vert v - w \Vert = \Vert v + w \Vert &\Leftrightarrow \Vert v - w \Vert^2 = \Vert v + w \Vert^2 \\
&\Leftrightarrow (v - w)^2 = (v + w)^2 \\
&\Leftrightarrow v^2 - 2 v \cdot w + w^2 = v^2 + 2 v \cdot w + w^2 \\
&\Leftrightarrow 4 v \cdot w = 0 \\
&\Leftrightarrow v \cdot w = 0 \\
\end{aligned}
$$

This is the desired justification.

The following two identities follow directly from the definition of the length.

### The Pythagoras theorem 

If $v, w$ are orthogonal, then

$$
\Vert v + w \Vert^2 = \Vert v \Vert^2 + \Vert w \Vert^2
$$

#### Proof

By the definition of norm, we have

$$
\Vert v + w \Vert = \sqrt{\langle v + w, v + w \rangle}
$$

Then by **SP 2**,

$$
\sqrt{\langle v + w, v + w \rangle} = \sqrt{\langle v + w, v \rangle + \langle v + w, w \rangle}
$$

Then by **SP 1**,

$$
\sqrt{\langle v + w, v \rangle + \langle v + w, w \rangle} = \sqrt{\langle v, v + w \rangle + \langle w, v + w \rangle}
$$

Then by **SP 2** again,

$$
\begin{aligned}
\sqrt{\langle v, v + w \rangle + \langle w, v + w \rangle} &= \sqrt{(\langle v, v \rangle + \langle v, w \rangle) + (\langle w, v \rangle + \langle w, w \rangle)} \\
&= \sqrt{\langle v, v \rangle + (\langle v, w \rangle + \langle w, v \rangle) + \langle w, w \rangle}
\end{aligned}
$$

And by **SP 1** again,

$$
\begin{aligned}
\sqrt{\langle v, v \rangle + (\langle v, w \rangle + \langle w, v \rangle) + \langle w, w \rangle} &= \sqrt{\langle v, v \rangle + (\langle v, w \rangle + \langle v, v \rangle) + \langle w, w \rangle} \\
&= \sqrt{\langle v, v \rangle + 2 \langle v, w \rangle + \langle w, w \rangle} \\
\end{aligned}
$$

And since $v, w$ are orthogonal, we have

$$
\begin{aligned}
\sqrt{\langle v, v \rangle + 2 \langle v, w \rangle + \langle w, w \rangle} &= \sqrt{\langle v, v \rangle + 2 \cdot 0 + \langle w, w \rangle} \\
&= \sqrt{\langle v, v \rangle + \langle w, w \rangle} \\
\end{aligned}
$$

Thus

$$
\Vert v + w \Vert = \sqrt{\langle v, v \rangle + \langle w, w \rangle}
$$

Hence

$$
\begin{aligned}
\Vert v + w \Vert^2 &= (\sqrt{\langle v, v \rangle + \langle w, w \rangle})^2 \\
&= \langle v, v \rangle + \langle w, w \rangle \\
&= \Vert v \Vert^2 + \Vert w \Vert^2
\end{aligned}
$$

Q.E.D.

###  The parallelogram law

For any $v, w$ we have

$$
\Vert v + w \Vert^2 + \Vert v - w \Vert^2 = 2 \Vert v \Vert ^2 + 2 \Vert w \Vert^2
$$

#### Proof

By the definition of norm, we have

$$
\begin{aligned}
\Vert v + w \Vert^2 + \Vert v - w \Vert^2 &= (\sqrt{\langle v + w, v + w \rangle})^2 + (\sqrt{\langle v - w, v - w \rangle})^2 \\
&= \langle v + w, v + w \rangle + \langle v - w, v - w \rangle
\end{aligned}
$$

Then by **SP 2**, we have

$$
\langle v + w, v + w \rangle + \langle v - w, v - w \rangle = (\langle v + w, v \rangle + \langle v + w, w \rangle) + (\langle v - w, v \rangle - \langle v - w, w \rangle) \\
$$

Then by **SP 1**, we have

$$
(\langle v + w, v \rangle + \langle v + w, w \rangle) + (\langle v - w, v \rangle - \langle v - w, w \rangle) = (\langle v, v + w \rangle + \langle w, v + w \rangle) + (\langle v, v - w \rangle - \langle w, v - w \rangle)
$$

Then by **SP 2** again, we have

$$
\begin{aligned}
(\langle v, v + w \rangle + \langle w, v + w \rangle) + (\langle v, v - w \rangle - \langle w, v - w \rangle) &= 
((\langle v, v \rangle + \langle v, w \rangle) + (\langle w, v \rangle + \langle w, w \rangle)) + ((\langle v, v \rangle - \langle v, w \rangle) - (\langle w, v \rangle - \langle w, w \rangle)) \\
&= 2 \langle v, v \rangle + (\langle v, w \rangle + \langle w, v \rangle - \langle v, w \rangle - \langle w, v \rangle) + 2 \langle w, w \rangle \\
&= 2 \langle v, v \rangle + 2 \langle w, w \rangle \\
&= 2 (\sqrt{\langle v, v \rangle})^2 + 2 (\sqrt{\langle w, w \rangle})^2 \\
&= 2 \Vert v \Vert^2 + 2 \Vert w \Vert^2
\end{aligned}
$$

Q.E.D.

## Projection of a vector along another one

Let $w$ be a vector of V such that $\Vert w \Vert \neq 0$.

For any $v$, there exists a unique number $c$ such that $v - c w$ is orthogonal to $w$. 

Indeed, for $v - cw$ to be orthogonal to $w$ we must have

$$
\langle v - c w, w \rangle = 0
$$

Then by **SP 1**,

$$
\langle v - c w, w \rangle = \langle w, v - c w \rangle
$$

Then by **SP 2**,

$$
\langle w, v - c w \rangle = \langle w, v \rangle - \langle w, c w \rangle
$$

Then by **SP 3**,

$$
\langle w, v \rangle - \langle w, c w \rangle = \langle w, v \rangle - c \langle w, w \rangle
$$

Then we have

$$
\langle w, v \rangle - c \langle w, w \rangle = 0
$$

Thus

$$
c = \frac{\langle w, v \rangle}{\langle w, w \rangle}
$$

And by $SP 1$, we have

$$
c = \frac{\langle v, w \rangle}{\langle w, w \rangle}
$$

Conversely, letting $c = \frac{\langle v, w \rangle}{\langle w, w \rangle}$ shows that $v - c w$ is orthogonal to $w$.

We call $c$ the **component of $v$ along $w$**.

We call $c w$ the **projection of $v$ along $w$**.

As with the case of $n$-space, we define the projection of $v$ along $w$ to be the vector $c w$, because of our usual picture:

![figure 2](figure_2.png)

In particular, if $w$ is a unit vector, then the component of $v$ along $w$ is simply

$$
c = \langle v, w \rangle
$$

#### Example 4

Let $V = R^n$ with the usual scalar product, i.e. the dot product. If $E^i$ is the $i$-th unit vector, and $X = (x_1, ... ,x_n)$ then the component of $X$ along $E^i$ is simply

$$
X \cdot E^i = x_i
$$

that is, the $i$-th component of $X$.

#### Example 5: Fourier coefficient

Let $V$ be the space of continuous functions on $[ - \pi, \pi]$. Let $f$ be the function given by $f(x) = \sin kx$, where $k$ is some integer $> 0$. Then

$$
\begin{aligned}
\Vert f \Vert = \sqrt{\langle f, f \rangle} &= (\int_{- \pi}^\pi (\sin k x)^2 d x)^\frac{1}{2} \\
&= ((\frac{x}{2} - \frac{\sin(2kx)}{4k})|_{x = \pi} - (\frac{x}{2} - \frac{\sin(2kx)}{4k})|_{x = - \pi})^\frac{1}{2} \\
&= ((\frac{\pi}{2} - \frac{\sin(2k \pi)}{4k}) - (\frac{- \pi}{2} - \frac{\sin(2k (- \pi))}{4k}))^\frac{1}{2} \\
&= ((\frac{\pi}{2} - \frac{\sin(2k \pi)}{4k}) - (- \frac{\pi}{2} + \frac{\sin(2k (- \pi))}{4k}))^\frac{1}{2} \\
&= \sqrt{\pi}
\end{aligned}
$$

In the present example of a vector space of functions, the component of $g$ along $f$ is called the **Fourier coefficient** of $g$ with respect to $f$. If $g$ is any continuous function on $[- \pi, \pi]$, then the Fourier coefficient of $g$ with respect to $f$ is

$$
\frac{\langle g, f \rangle}{\langle f, f \rangle} = \frac{\int_{- \pi}^\pi g(x) \sin k x d x}{\int_{- \pi}^\pi (\sin k x)^2 d x} = \frac{1}{\pi} \int_{- \pi}^\pi g(x) \sin k x d x
$$

## Theorem 1.1: Schwarz inequality

For all $v, w \in V$, we have

$$
\vert \langle v, w \rangle \vert \leq \Vert v \Vert \Vert w \Vert
$$

### Proof

#### If $w = \mathit{0}$

If $w = \mathit{0}$, by [lemma 1.0](#Lemma-1.0), we have

$$
\vert \langle v, w \rangle \vert = \vert \langle v, \mathit{0} \rangle \vert = 0
$$

And by lemma 1.0 again, we have

$$
\Vert v \Vert \Vert w \Vert = \Vert v \Vert \Vert \mathit{0} \Vert = \Vert v \Vert \sqrt{\langle \mathit{0}, \mathit{0} \rangle} = \Vert v \Vert \cdot 0 = 0
$$

Hence we have $\vert \langle v, w \rangle \vert = \Vert v \Vert \Vert w \Vert$. Thus $\vert \langle v, w \rangle \vert \leq \Vert v \Vert \Vert w \Vert$.

#### If $w \neq \mathit{0}$

Let $e$ is a unit vector, then we have $\Vert e \Vert = 1$.

If $c$ is the component of $v$ along $e$, then $v - c e$ is orthogonal to $e$.

And by **SP 3**, $v - c e$ is also orthogonal to $c e$.

Then by the [Pythagoras theorem](#The-Pythagoras-theorem), we have

$$
\Vert v \Vert^2 = \Vert (v - ce) + c e \Vert^2 = \Vert v - ce \Vert^2 + \Vert ce \Vert^2 \\
$$

And since $e$ is a unit vector, we have

$$
\Vert ce \Vert^2 = c^2 \Vert e \Vert^2 = c^2
$$

Thus

$$
\Vert v - ce \Vert^2 + \Vert ce \Vert^2 = \Vert v - ce \Vert^2 + c^2
$$

Thus

$$
\Vert v \Vert^2 = \Vert v - ce \Vert^2 + c^2
$$

Hence $c^2 \leq \Vert v \Vert^2$. Hence $\vert c \vert \leq \Vert v \Vert$ (\*).

Finally, if $w$ is arbitrary vector in $V$ that $\neq \mathit{0}$, then $e = \frac{w}{\Vert w \Vert}$ is a unit vector, so that we have

$$
c = \langle v, e \rangle = \langle v, \frac{w}{\Vert w \Vert} \rangle
$$

Then by (\*), we have

$$
\vert \langle v, \frac{w}{\Vert w \Vert} \rangle \vert \leq \Vert v \Vert
$$

By **SP 3**, we have

$$
\langle v, \frac{w}{\Vert w \Vert} \rangle = \langle v, \frac{1}{\Vert w \Vert} w \rangle = \frac{1}{\Vert w \Vert}  \langle v, w \rangle
$$

Then we have

$$
\vert \langle v, w \rangle \vert \leq \Vert v \Vert \Vert w \Vert
$$

Q.E.D.

## Theorem 1.2: Triangle inequality

If $v, w \in V$, then

$$
\Vert v + w \Vert \leq \Vert v \Vert + \Vert w \Vert
$$

### Proof

We have

$$
\Vert v + w \Vert^2 = (\sqrt{\langle v + w, v + w \rangle})^2 = \langle v + w, v + w \rangle
$$

By **SP 1 and SP 2**, we have

$$
\begin{aligned}
\langle v + w, v + w \rangle &= \langle v + w, v \rangle + \langle v + w, w \rangle \\
&= \langle v, v + w \rangle + \langle w, v + w \rangle \\
&= \langle v, v \rangle + \langle v, w \rangle + \langle w, v \rangle + \langle w, w \rangle \\
&= \langle v, v \rangle + 2 \langle v, w \rangle + \langle w, w \rangle
\end{aligned}
$$

Thus

$$
\Vert v + w \Vert^2 = \langle v, v \rangle + 2 \langle v, w \rangle + \langle w, w \rangle
$$

And by [Schwartz inequality](#Theorem-1.1:-Schwarz-inequality), we have

$$
\begin{array}{l}
\langle v, v \rangle &\leq \vert \langle v, v \rangle \vert &\leq \Vert v \Vert^2 \\
\langle v, w \rangle &\leq \vert \langle v, w \rangle \vert &\leq \Vert v \Vert \Vert w \Vert \\
\langle w, w \rangle &\leq \vert \langle w, w \rangle \vert &\leq \Vert w \Vert^2 \\
\end{array}
$$

Thus we have

$$
\begin{aligned}
\Vert v + w \Vert^2 &= \langle v, v \rangle + 2 \langle v, w \rangle + \langle w, w \rangle \\
&\leq \Vert v \Vert^2 + 2 \Vert v \Vert \Vert w \Vert + \Vert w \Vert^2 \\
&= (\Vert v \Vert + \Vert w \Vert)^2
\end{aligned}
$$

And since we are in the positive definite case, we have

$$
\begin{array}{c}
\Vert v + w \Vert &\geq 0 \\
\Vert v \Vert + \Vert w \Vert &\geq 0
\end{array}
$$

Thus

$$
\Vert v + w \Vert^2 \leq (\Vert v \Vert + \Vert w \Vert)^2
$$

is equivalent to

$$
\Vert v + w \Vert \leq \Vert v \Vert + \Vert w \Vert
$$

Thus we have proven $\Vert v + w \Vert \leq \Vert v \Vert + \Vert w \Vert$. Q.E.D.

### Mutual orthogonal set and its component along the elements of the set

Let $v_1, ... ,v_n$ be non-zero elements of $V$ which are mutually orthogonal, that is $\langle v_i, v_j \rangle$ if $i \neq j$. Let $v$ be any vector in $V$. Let $c_i$ be the component of $v$ along $v_i$. Then

$$
v - c_1 v_1 - ... - c_n v_n
$$

is orthogonal to $v_1, ..., v_n$.

#### Proof

Since $c_i$ is the component of $v$ along $v_i$, we have

$$
c_i = \frac{\langle v, v_i \rangle}{\langle v_i, v_i \rangle}
$$


Then for any $i$, we have

$$
\begin{aligned}
\langle v_i, v - c_1 v_1 - ... - c_n v_n \rangle &= \langle v_i, v \rangle - c_1 \langle v_i, v_1 \rangle - ... - c_n \langle v_i, v_n \rangle \\
\end{aligned}
$$

$\langle v_i, v_j \rangle$ if $i \neq j$ for $v_i, v_j$ in $v_1, ..., v_n$, we have

$$
\begin{aligned}
\langle v_i, v - c_1 v_1 - ... - c_n v_n \rangle &= \langle v_i, v \rangle - c_1 \langle v_i, v_1 \rangle - ... - c_n \langle v_i, v_n \rangle \\
&= \langle v_i, v \rangle - c_i \langle v_i, v_i \rangle \\
&= \langle v_i, v \rangle - \frac{\langle v, v_i \rangle}{\langle v_i, v_i \rangle} \langle v_i, v_i \rangle \\
&= \langle v_i, v \rangle - \langle v, v_i \rangle \\
&= \langle v_i, v \rangle -  \langle v_i, v \rangle \\
&= 0
\end{aligned}
$$

Thus $v - c_1 v_1 - ... - c_n v_n$ is orthogonal to $v_i$ in $v_1, ..., v_n$. Q.E.D.

### Theorem 1.3

Let $v_1, ..., v_n$ be vectors which are mutually orthogonal, and such that $\Vert v_i \Vert \neq 0$ for all $i$. Let $v$ be an element of $V$, and let $c_i$ be the component of $v$ along $v_i$. Let $a_1, ..., a_n$ be numbers. Then

$$
\Vert v - \sum_{k = 1}^n c_k v_k \Vert \leq \Vert v - \sum_{k = 1}^n a_k v_k \Vert
$$

#### Proof

We [already know that](#Mutual-orthogonal-set-and-its-component-along-the-elements-of-the-set)

$$
v - \sum_{k = 1}^n c_k v_k
$$

is orthogonal to each $v_i$, $i = 1, ..., n$. Hence it is orthogonal to any linear combination of of $v_1, ..., v_n$.

Then by the [Pythagoras theorem](#The-Pythagoras-theorem), we have

$$
\Vert v - \sum_{k = 1}^n a_k v_k \Vert^2 = \Vert v - \sum_{k = 1}^n c_k v_k \Vert^2 + \Vert \sum_{k = 1}^n (c_k - a_k) v_k \Vert^2
$$

Thus

$$
\Vert v - \sum_{k = 1}^n c_k v_k \Vert^2 \leq \Vert v - \sum_{k = 1}^n a_k v_k \Vert^2
$$

And in the positive definite case, we have

$$
\Vert v - \sum_{k = 1}^n c_k v_k \Vert \leq \Vert v - \sum_{k = 1}^n a_k v_k \Vert
$$

Q.E.D.

### Theorem 1.4: Bessel inequality

If $v_1, ..., v_n$ are mutually orthogonal unit vectors, and if $c_i$ is the component of $v$ along $v_i$ then

$$
\sum_{i = 1}^n c_i^2 \leq \Vert v \Vert^2
$$

#### Proof

We [already know that](#Mutual-orthogonal-set-and-its-component-along-the-elements-of-the-set) $v_1, ..., v_n, \sum_{i = 1}^n c_i v_i$ are mutually orthogonal. Therefore

$$
\begin{aligned}
\Vert v \Vert^2 &= \Vert v - \sum_{i = 1}^n c_i v_i \Vert^2 + \Vert \sum_{i = 1}^n c_i v_i \Vert^2 & \text{by Pythagoras} \\
&\geq \Vert \sum_{i = 1}^n c_i v_i \Vert^2 &\text{by the definition of norm} \\
&= \sum_{i = 1}^n \Vert c_i v_i \Vert^2 &\text{by Pythagoras} \\
&= \sum_{i = 1}^n c_i^2 \Vert v_i \Vert^2 &\text{by SP 3} \\
&= \sum_{i = 1}^n c_i^2 &\text{as $v_1, ..., v_n$ are unit vectors}
\end{aligned}
$$

Q.E.D.

### Theorem 1.5

*This inspired by [Prof Robert Ghrist's lecture on basic vector algebra](https://youtu.be/3V_3dnO-0lo).*

Let $V$ be a vector space over $V$ where a positive definite scalar product is defined. Let $v$ be any vector in $V$. Then $\Vert v \Vert = 0$ if and only if $v = \mathit{0}$.

#### Proof

##### Proof: $\Vert v \Vert = 0 \Rightarrow v = \mathit{0}$

Since

$$
\Vert v \Vert = 0
$$

Then

$$
\sqrt{\langle v, v \rangle} = 0
$$

And since the scalar product is positive-definite, then we have $v = \mathit{0}$.

##### Proof: $v = \mathit{0} \Rightarrow \Vert v \Vert$

Since the scalar product is positive definite, we have $\Vert \mathit{0}, \mathit{0} \Vert = \sqrt{\langle \mathit{0}, \mathit{0} \rangle} = \sqrt{0} = 0$.

Q.E.D.
