## Definitions: Orthogonal bases and orthonormal bases

Let $V$ be a vector space with a positive definite scalar product throughout this section.

A basis $\{v_1, ..., v_n\}$ of V is said to be **orthogonal** if its elements are mutually orthogonal, i.e. $\langle v_i, v_j \rangle$ whenever $i \neq j$.

If in addition, each vector of the basis has norm $1$, then the basis is called **orthonormal**.

The standard unit vectors of $R^n$ form an orthonormal basis of $R^n$, with respect to the ordinary dot product.

### Theorem 2.1

Let $V$ be a finite dimensional vector space, with a positive definite scalar product. Let $W$ be a subspace of $V$, and let $\{w_1, ... ,w_m\}$ be an orthogonal basis of $W$ If $W \neq V$, then there exist elements $w_{m+1}, ..., w_n$ of $V$ such that $\{w_1, ..., w_n\}$ is an orthogonal basis of $V$.

#### Proof

The method of proof is as important as the theorem, and is called the **Gram-Schmidt orthogonalization process**.

Since $W$ is a subspace of $V$, then its $w_1, ..., w_m$ are also linear independent vectors in $V$.

And since $W \neq V$, by [corollary 3.5 in chapter 1.3](1_vector_spaces/1_vector_spaces.ipynb#Corollary-3.5.), we have $\dim W \leq \dim V$. Thus $m \leq n$.

Then by [corollary 3.6 in chapter 1.3](../1_vector_spaces/1_vector_spaces.ipynb#Corollary-3.6.), there are some vectors $v_{m + 1}, ..., v_n \in V$ such that

$$
\{w_1, ..., w_m, v_{m + 1}, ..., v_n \}
$$

is a basis of $V$.

Let $W_{m + 1}$ be the space generated by $w_1, ..., w_m, v_{m + 1}$. We shall obtain one of its basis by letting

$$
c_1 = \frac{\langle v_{m + 1}, w_1 \rangle}{\langle w_1, w_1 \rangle}, ..., c_m = \frac{\langle v_{m + 1}, w_m \rangle}{\langle w_m, w_m \rangle}
$$

and

$$
w_{m + 1} = v_{m + 1} - c_1 w_1 - ... - c_m w_m
$$

which [we know](../5_1_scalar_products.ipynb#Mutual-orthogonal-set-and-its-component-along-the-elements-of-the-set) $w_{m + 1}$ is orthogonal to $w_1, ..., w_m$. Furthermore, since $w_1, ..., w_m, v_{m + 1}$ are linearly independent, we have $w_{m + 1} \neq \mathit{0}$.

And since now we can write $v_{m + 1}$ as

$$
v_{m + 1} = w_{m + 1} + c_1 w_1 + ... + c_m w_m
$$

Then we know $v_{m + 1}$ is in the space generated by $w_1, ..., w_{m + 1}$.

Hence $\{w_1, ... ,w_{m + 1}\}$ is an orthogonal basis of $W_{m + 1}$ We can now proceed by induction, showing that the space $W_{m + s}$ generated by

$$
w_1, ..., w_m, v_{m + 1}, ..., v_{m + s}
$$

has an orthogonal basis

$$
\{w_1, ..., w_{m + 1}, ..., w_{m + s}\}
$$

with $s = 1, ..., n - m$. This concludes the proof. Q.E.D.

### Corollary 2.2

Let $V$ be a finite dimensional vector space with a positive definite scalar product. Assume that $V \neq \{\mathit{0}\}$. Then $V$ has an orthogonal basis.

#### Proof

Since $V \neq \{\mathit{0}\}$, and $V$ is a finite dimensional vector space, let $\{v_1, ...,v_n\}$ be an arbitrary basis of $V$.

Then we let

$$
\begin{array}{l}
v_1^\prime &= v_1 \\
v_2^\prime &= v_2 - \frac{\langle v_2, v_1^\prime \rangle}{\langle v_1^\prime, v_1^\prime \rangle} v_1 \\
v_3^\prime &= v_3 - \frac{\langle v_3, v_2^\prime \rangle}{\langle v_2^\prime, v_2^\prime \rangle} v_2 - \frac{\langle v_3, v_1^\prime \rangle}{\langle v_1^\prime, v_1^\prime \rangle} v_1 \\
\vdots \\
v_n^\prime &= v_n - \frac{\langle v_n, v_{n - 1}^\prime \rangle}{\langle v_{n - 1}^\prime, v_{n - 1}^\prime \rangle} v_{n - 1} - ... - \frac{\langle v_n, v_1^\prime \rangle}{\langle v_1^\prime, v_1^\prime \rangle} v_1 \\
\end{array}
$$

Then ${v_1^\prime ... ,v_n^\prime}$ is an orthogonal basis. Q.E.D.

Given an orthogonal basis, we can always obtain an orthonormal basis by dividing each vector by its norm.
