## Definitions: Orthogonal bases and orthonormal bases

Let $V$ be a vector space with a positive definite scalar product throughout this section.

A basis $\{v_1, ..., v_n\}$ of V is said to be **orthogonal** if its elements are mutually orthogonal, i.e. $\langle v_i, v_j \rangle$ whenever $i \neq j$.

If in addition, each vector of the basis has norm $1$, then the basis is called **orthonormal**.

The standard unit vectors of $R^n$ form an orthonormal basis of $R^n$, with respect to the ordinary dot product.

### Theorem 2.1

Let $V$ be a finite dimensional vector space, with a positive definite scalar product. Let $W$ be a subspace of $V$, and let $\{w_1, ... ,w_m\}$ be an orthogonal basis of $W$ If $W \neq V$, then there exist elements $w_{m+1}, ..., w_n$ of $V$ such that $\{w_1, ..., w_n\}$ is an orthogonal basis of $V$.

#### Proof

The method of proof is as important as the theorem, and is called the **Gram-Schmidt orthogonalization process**.

Since $W$ is a subspace of $V$, then its $w_1, ..., w_m$ are also linear independent vectors in $V$.

And since $W \neq V$, by [corollary 3.5 in chapter 1.3](1_vector_spaces/1_vector_spaces.ipynb#Corollary-3.5.), we have $\dim W \leq \dim V$. Thus $m \leq n$.

Then by [corollary 3.6 in chapter 1.3](../1_vector_spaces/1_vector_spaces.ipynb#Corollary-3.6.), there are some vectors $v_{m + 1}, ..., v_n \in V$ such that

$$
\{w_1, ..., w_m, v_{m + 1}, ..., v_n \}
$$

is a basis of $V$.

Let $W_{m + 1}$ be the space generated by $w_1, ..., w_m, v_{m + 1}$. We shall obtain one of its basis by letting

$$
c_1 = \frac{\langle v_{m + 1}, w_1 \rangle}{\langle w_1, w_1 \rangle}, ..., c_m = \frac{\langle v_{m + 1}, w_m \rangle}{\langle w_m, w_m \rangle}
$$

and

$$
w_{m + 1} = v_{m + 1} - c_1 w_1 - ... - c_m w_m
$$

which [we know](../5_1_scalar_products.ipynb#Mutual-orthogonal-set-and-its-component-along-the-elements-of-the-set) $w_{m + 1}$ is orthogonal to $w_1, ..., w_m$. Furthermore, since $w_1, ..., w_m, v_{m + 1}$ are linearly independent, we have $w_{m + 1} \neq \mathit{0}$.

And since now we can write $v_{m + 1}$ as

$$
v_{m + 1} = w_{m + 1} + c_1 w_1 + ... + c_m w_m
$$

Then we know $v_{m + 1}$ is in the space generated by $w_1, ..., w_{m + 1}$.

Hence $\{w_1, ... ,w_{m + 1}\}$ is an orthogonal basis of $W_{m + 1}$ We can now proceed by induction, showing that the space $W_{m + s}$ generated by

$$
w_1, ..., w_m, v_{m + 1}, ..., v_{m + s}
$$

has an orthogonal basis

$$
\{w_1, ..., w_{m + 1}, ..., w_{m + s}\}
$$

with $s = 1, ..., n - m$. This concludes the proof. Q.E.D.

### Corollary 2.2

Let $V$ be a finite dimensional vector space with a positive definite scalar product. Assume that $V \neq \{\mathit{0}\}$. Then $V$ has an orthogonal basis.

#### Proof

Since $V \neq \{\mathit{0}\}$, and $V$ is a finite dimensional vector space, let $\{v_1, ...,v_n\}$ be an arbitrary basis of $V$.

Let $W_1$ be the space generated by $v_1$. We know $\{v_1\}$ is an orthogonal basis of $W_1$.

Then following the Gram-Schmidt orthogonalization process, we let

$$
\begin{array}{l}
v_1^\prime &= v_1 \\
v_2^\prime &= v_2 - \frac{\langle v_2, v_1^\prime \rangle}{\langle v_1^\prime, v_1^\prime \rangle} v_1 \\
v_3^\prime &= v_3 - \frac{\langle v_3, v_2^\prime \rangle}{\langle v_2^\prime, v_2^\prime \rangle} v_2 - \frac{\langle v_3, v_1^\prime \rangle}{\langle v_1^\prime, v_1^\prime \rangle} v_1 \\
\vdots \\
v_i^\prime &= v_i - \frac{\langle v_i, v_{i - 1}^\prime \rangle}{\langle v_{i - 1}^\prime, v_{i - 1}^\prime \rangle} v_{i - 1} - ... - \frac{\langle v_i, v_1^\prime \rangle}{\langle v_1^\prime, v_1^\prime \rangle} v_1 \\
\vdots \\
v_n^\prime &= v_n - \frac{\langle v_n, v_{n - 1}^\prime \rangle}{\langle v_{n - 1}^\prime, v_{n - 1}^\prime \rangle} v_{n - 1} - ... - \frac{\langle v_n, v_1^\prime \rangle}{\langle v_1^\prime, v_1^\prime \rangle} v_1 \\
\end{array}
$$

Then ${v_1^\prime ... ,v_n^\prime}$ is an orthogonal basis. Q.E.D.

We can write 
$$
v_i^\prime = v_i - \sum_{j = 0}^{i - 1} \frac{\langle v_i, v_j^\prime \rangle}{\langle v_j^\prime, v_j^\prime \rangle} v_j
$$

Given an orthogonal basis, we can always obtain an orthonormal basis by dividing each vector by its norm.

### Example 1

Find an orthonormal basis for the vector space generated by the vectors $\begin{pmatrix} 1 \\ 1 \\ 0 \\ 1 \end{pmatrix}$, $\begin{pmatrix} 1 \\ -2 \\ 0 \\0 \end{pmatrix}$, and $\begin{pmatrix} 1 \\ 0 \\ -1 \\ 2 \end{pmatrix}$.

```python
import numpy as np

def gram_schmidt(vectors):
    orthogonalized = []
    for i, v in enumerate(vectors):
        terms = [np.dot(v, v_j_prime) / np.dot(v_j_prime, v_j_prime) * v_j_prime for v_j_prime in orthogonalized]
        orthogonalized.append(v - sum(terms))
    return orthogonalized

def normalized(vectors):
    return [v / np.linalg.norm(v) for v in vectors]

def orthonormalize_gram_schmidt(vectors):
    return normalized(gram_schmidt(vectors))
```

```python
A = np.array([1, 1, 0, 1]).T
B = np.array([1, -2, 0, 0]).T
C = np.array([1, 0, -1, 2]).T

ex_1_solution = orthonormalize_gram_schmidt([A, B, C])
ex_1_solution
```

```python
## Verify for orthogonality
from itertools import combinations

def assert_orthogonality(vectors):
    np.testing.assert_almost_equal([0, 0, 0], [np.dot(v_1, v_2) for v_1, v_2 in combinations(vectors, 2)])

assert_orthogonality(ex_1_solution)
```

### Theorem 2.3

Let $V$ be a vector space over $R$ with a positive definite scalar product, of dimension $n$. Let $W$ be a subspace of $V$ of dimension $r$. Let $W^\perp$ be the subspace of $V$ consisting of all elements which are orthogonal to $W$. Then $V$ is the direct sum of $W$ and $W^\perp$, and $W^\perp$ has dimension $n - r$. In other words,

$$
\dim W + \dim W^\perp = \dim V
$$

#### Proof

##### If $W = \{\mathit{0}\}$

If $W = \{\mathit{0}\}$, since $\mathit{0}$ is orthogonal to any vector in $V$ including itself, then we have $W^\perp = V$. Thus

$$
\dim W + \dim W^\perp = 0 + \dim V = \dim V
$$

It is obvious that $V = W + W^\perp$ and $W \cap W^\perp = \{\mathit{0}\}$. By [theorem 4.1 in chapter 1.4](1_vector_spaces/1_vector_spaces.ipynb#Theorem-4.1.), $V = W \oplus W^\perp$.

#####  If $W = V$

If $W = V$, then $W^\perp = \{\mathit{0}\}$. Thus

$$
\dim W + \dim W^\perp = \dim V + 0 = \dim V
$$

It is obvious that $V = W + W^\perp$ and $W \cap W^\perp = \{\mathit{0}\}$. By [theorem 4.1 in chapter 1.4](1_vector_spaces/1_vector_spaces.ipynb#Theorem-4.1.), $V = W \oplus W^\perp$.


##### If $W \neq \{\mathit{0}\}$ and $W \neq V$

Assume $W \neq \{\mathit{0}\}$ and $W \neq V$.

Since $V$ has a positive definite scalar product, then $W$ as a subspace has the same positive definite scalar product.

Then by [corollary 2.2](#Corollary-2.2), $W$ has an orthonormal basis $\{w_1, ..., w_r\}$.

Then by [theorem 2.1](#Theorem-2.1), there exists vectors $u_{r + 1}, ..., u_n$ of $V$ such that

$$
\{w_1, ..., w_r, u_{r + 1}, ..., u_n\}
$$

is an orthonormal basis of $V$. We now contend that $\{u_{r + 1}... , u_n\}$ is an orthonormal basis of $W^\perp$.

Let $u$ be an element of $W^\perp$. Recall that $\{w_1, ..., w_r\}$ is an orthonormal basis of $V$. Then there exists numbers $x_1, ..., x_n$ such that

$$
u = x_1 w_1 + ... + x_r w_r + x_{r + 1} w_{r + 1} + ... + x_n w_n
$$

And since $u \in W^\perp$, thus $u$ is orthogonal to $W$. Recall that $\{w_1, ..., w_r\}$ is an orthonormal basis of $W$. Then for $i$ in $1, ..., r$, we have

$$
\begin{aligned}
0 &= \langle u, w_i \rangle = \langle x_1 w_1 + ... + x_r w_r + x_{r + 1} w_{r + 1} + ... + x_n w_n, w_i \rangle \\
&= x_1 \langle w_1, w_i \rangle + ... + x_{i - 1} \langle w_{i - 1}, w_i \rangle + x_i \langle w_i, w_i \rangle + x_{i + 1} \langle w_{i + 1}, w_i \rangle +... + x_n \langle w_n, w_i \rangle \\
&= x_1 \cdot 0 + ... + x_{i - 1} \cdot 0 + x_i \langle w_i, w_i \rangle + x_{i + 1} \cdot 0 + ... + x_n w_n \\
&= x_i \langle w_i, w_i \rangle \\
&= x_i
\end{aligned}
$$

In other words, $x_i = 0$ for $i = 1, ..., r$. Therefore, for any $u \in W\perp$, $u$ is a linear combination of $u_{r + 1}, ..., u_n$.  **(1)**.

Conversely, let $x_{r + 1}, ..., x_n$ be any numbers, then let $u = x_{r + 1} u_{r + 1} + ... + x_n u_n$, which is a linear combination of $u_{r + 1}, ..., u_n$.

Then for $w_i$ of $i = 1, ..., r$,

$$
\begin{aligned}
\langle w_i, u \rangle &= \langle w_i, x_{r + 1} u_{r + 1} + ... + x_n u_n \rangle \\
&= x_{r + 1} \langle w_i, u_{r + 1} \rangle + ... + x_n \langle w_i, u_n \rangle &\text{By SP 2 and SP 3} \\
&= x_{r + 1} \cdot 0 + ... + x_n \cdot 0 &\text{As $\{w_1, ..., w_r, u_{r + 1}, ..., u_n\}$ is an orthonormal basis of $V$} \\
&= 0
\end{aligned}
$$

Hence $u$ is orthogonal to $w_i$ of $i = 1, ..., r$. By [definition](../5_1_scalar_products.ipynb#Definition:-Orthogonal-space), $u \perp W$. **(2)**.

Hence by **(1)** and **(2)**, $u_{r + 1}, ..., u_n$ generates $W^\perp$.

Recall again $\{w_1, ..., w_r, u_{r + 1}, ..., u_n\}$ is an orthonormal basis of $V$. Thus
- $u_{r + 1}, ..., u_n$ are linearly independent.
- $u_{r + 1}, ..., u_n$ are mutually orthogonal, and are all of norm $1$.

Thus $\{u_{r + 1}, ..., u_n\}$ is an orthonormal basis of $W^\perp$.

Hence $\dim V = \dim W + \dim W^\perp$.

And $V = W + W\perp$. And $W \cap W^\perp = \{0 \cdot w_1 + ... + 0 \cdot w_r + 0 \cdot u_{r + 1} + ... + 0 \cdot u_n\} = \{\mathit{0}\}$. By [theorem 4.1 in chapter 1.4](1_vector_spaces/1_vector_spaces.ipynb#Theorem-4.1.), $V = W \oplus W^\perp$.

Q.E.D.

The space $W^\perp$ is called the orthogonal complement of $W$.

### Positive definitive scalar products and orthonormal base

Let $V$ be a finite dimensional vector space over $R$, with a positive definite scalar product. Let $\{e_l , ... , e_n\}$ be an orthonormal basis. Let $v, w \in V$. There exist numbers $x_1, ..., x_n \in R$ and $y_1, ..., y_n \in R$ such that

$$
\begin{aligned}
v = x_1 e_1 + ... + x_n e_n && \text{and} && w = y_1 e_1 + ... + y_n e_n
\end{aligned}
$$

Then

$$
\begin{aligned}
\langle v, w \rangle &= \langle v, y_1 e_1 + ... + y_n e_n \rangle \\
&= y_1 \langle v, e_1 \rangle + ... + y_n \langle v, e_n \rangle \\
&= y_1 \langle e_1, v \rangle + ... + y_n \langle e_n, v \rangle &\text{SP 2 and SP 3} \\
&= y_1 \langle e_1, x_1 e_1 + ... + x_n e_n \rangle + ... + y_n \langle e_n, x_1 e_1 + ... + x_n e_n \rangle \\
&= y_1 (x_1 \langle e_1, e_1 \rangle + ... + x_n \langle e_1, e_n \rangle) + ... + y_n (x_1 \langle e_n, e_1 \rangle + ... + x_n \langle e_n, e_n \rangle) &\text{SP 2 and SP 3} \\
&= y_1 (x_1 \langle e_1, e_1 \rangle + ... + x_n \cdot 0) + ... + y_n (x_1 \cdot 0 + ... + x_n \langle e_n, e_n \rangle) &\text{As $e_l , ... , e_n$ are mutually orthogonal} \\
&= x_1 y_1 \langle e_1, e_1 \rangle + ... + x_n y_n  \langle e_n, e_n \rangle \\
&= x_1 y_1 + ... + x_n y_n &\text{As $\{e_l , ... , e_n\}$ is an orthonormal basis thus $\langle e_i, e_i \rangle = 1$} \\
&= \sum_{i = 1}^n x_i y_i
\end{aligned}
$$

Thus in terms of this orthonormal basis, if $X, Y$ are the coordinate vectors of $v$ and $w$ respectively, the scalar product is given by the ordinary dot product $X \cdot Y$ of the coordinate vectors.

This is definitely not the case if we deal with a basis which is not orthonormal. Let $\{v_1, ..., v_m\}$ be an arbitrary basis of $V$, then we write

$$
\begin{aligned}
v &= x_1 v_1 + ... + x_n v_n \\
w &= y_1 v_1 + ... + y_n v_n
\end{aligned}
$$

in terms of the basis, then

$$
\langle v, w \rangle = \sum_{i = 1}^n \sum_{j = 1}^n x_i y_j \langle v_i, v_j \rangle
$$

Each $\langle v_i, v_j \rangle$ is a number. If we let $a_{i j} = \langle v_i, v_j \rangle$, then

$$
\langle v, w \rangle = \sum_{i = 1}^n \sum_{j = 1}^n a_{i j} x_i y_j
$$

### Hermitian products

We shall now describe the modification necessary to adapt the preceding results to vector spaces over the complex numbers.

We wish to preserve the notion of a positive definite scalar product as far as possible. Since the dot product of vectors with complex coordinates may be equal to $0$ without the vectors being equal to $\mathit{0}$, we must change something in the definition.

It turns out that the needed change is very slight.

Let $V$ be a vector space over the complex numbers. A hermitian product on $V$ is a rule which to any pair of elements $v, w$ of $V$ associates a complex number, denoted again by $\langle v, w \rangle$, satisfying the following properties

- **HP 1.** $\langle v, w \rangle = \overline{\langle w, v \rangle}$ for $v, w \in V$. (Here the bar denotes complex conjugate.)
- **HP 2.** If $u, v, w$ are elements of $V$, then
$$
\langle u, v + w \rangle = \langle u, v \rangle + \langle u, w \rangle 
$$
- **HP 3.** If $\alpha \in C$, then
$$
\begin{aligned}
\langle \alpha u, v \rangle &= \alpha \langle u, v \rangle &&\text{and}&& \langle \alpha u, v \rangle = \overline{
\alpha} \langle u, v \rangle
\end{aligned}
$$

The hermitian product is said to be **positive definite** if
- $\langle v, v \rangle \geq 0$ for all $v \in V$
- and $\langle v, v \rangle \geq 0$ if $v \neq \mathit{0}$

We define the words **orthogonal**, **perpendicular**, **orthogonal basis**, **orthogonal complement** as before. There is nothing to change either in our definition of **component** and **projection of v along w**, or in the remarks which we made concerning these.

### Example 3

Let $V = C^n$. If $X = \begin{pmatrix} x_1 \\ \vdots \\ x_n \end{pmatrix}$ and $Y = \begin{pmatrix} y_1 \\ \vdots \\ y_n \end{pmatrix}$ are vectors in $C^n$, we define their **hermitian product** to be

$$
\langle X, Y \rangle = x_1 \overline{y_1} + ... + x_n \overline{y_n}
$$

#### Verifying HP 1

$$
\begin{aligned}
\overline{\langle Y, X \rangle} &= \overline{y_1 \overline{x_1} + ... + y_n \overline{x_n}} \\
&= \overline{y_1 \overline{x_1}} + ... + \overline{y_n \overline{x_n}} \\
&= \overline{y_1} x_1 + ... +  \overline{y_n} x_n \\
&= x_1 \overline{y_1} + ... + x_n \overline{y_n}
\end{aligned}
$$

#### Verifying HP 2

Let $Z = \begin{pmatrix} z_1 \\ \vdots \\ z_n \end{pmatrix}$ be a vector in $C^n$. Then

$$
\begin{aligned}
\langle X, Y + Z \rangle &= x_1 (\overline{y_1 + z_1}) + ... + x_n (\overline{y_n + z_n}) \\
&= x_1 (\overline{y_1} + \overline{z_1}) + ... + x_n (\overline{y_n} + \overline{z_n}) \\
&= (x_1 \overline{y_1} + ... + x_1 \overline{y_n}) + (x_1 \overline{z_1} + ... + x_n \overline{z_n}) \\
&= \langle X, Y \rangle + \langle X, Z \rangle
\end{aligned}
$$

#### Verifying HP 3

Let $\alpha$ be any number in $C$. Then we have

$$
\begin{aligned}
\langle \alpha X, Y \rangle &= \alpha x_1 \overline{y_1} + ... + \alpha x_n \overline{y_n} \\
&= \alpha (x_1 \overline{y_1} + ... + x_n \overline{y_n}) \\
&= \alpha \langle X, Y \rangle
\end{aligned}
$$

$$
\begin{aligned}
\langle X, \alpha Y \rangle &= x_1 (\alpha \overline{y_1}) + ... + x_n (\alpha \overline{y_n}) \\
&= \alpha (x_1 \overline{y_1} + ... + x_n \overline{y_n}) \\
&= \alpha \langle X, Y \rangle
\end{aligned}
$$

#### Verifying positive definiteness

If $X \neq \mathit{0}$, then there are some $x_i \neq 0$, and thus $x_i \overline{x_i} > 0$. Hence $\langle X, X \rangle > 0$.

#### Positive definiteness does not hold for dot products of complex numbers

In the case of $n = 2$ thus in $C^2$, let $X = \begin{pmatrix} 1 \\ i \end{pmatrix}$, then

$$
X \cdot X = 1 + (-1) = 0
$$

### Norm in vector spaces over the complex field

We now come to the theory of the norm. Let V be a vector space over $C$, with a positive definite hermitian product. If $v \in V$, we define its norm by letting

$$
\Vert v \Vert = \sqrt{\langle v, v \rangle}
$$

Since $\langle v, v \rangle$ is real, $\geq 0$, its square root is taken as usual to be the unique real number $\geq 0$ whose square is $\langle v, v \rangle$.

#### Schwarz inequality

We have the Schwarz inequality, namely

$$
\vert \langle v, w \rangle \vert \leq \Vert v \Vert \Vert w \Vert
$$

The three properties of the norm hold as in the real case:

1. For all $v \in V$, we have $\Vert v \Vert > 0$, and $= 0$ *if and only if* $v = \mathit{0}$.

##### Proof

Let $v = \begin{pmatrix}v_1 \\ \vdots \\ v_n \end{pmatrix}$


Then

$$
\begin{aligned}
\Vert v \Vert &= \sqrt{\langle v, v \rangle} \\
&= \sqrt{v_1 \overline{v_1} + ... + v_n \overline{v_n}} \\
&= \sqrt{\vert v_1 \vert^2 + ... + \vert v_n \vert^2}
\end{aligned}
$$

Thus $\Vert v \Vert$ if and only if $v = \mathit{0}$.

2. For any complex number $\alpha$, we have $\Vert \alpha v \Vert = \vert \alpha \vert \Vert v \Vert$

##### Proof

$$
\begin{aligned}
\Vert \alpha v \Vert &= \sqrt{\alpha v_1 \overline{\alpha v_1} + ... + \alpha v_1 \overline{\alpha v_1}} \\
&= \sqrt{\alpha^2 (v_1 \overline{v_1} + ... + v_n \overline{v_n})} \\
&= \vert \alpha \vert \Vert v \Vert
\end{aligned}
$$

3. (Triangle inequality) For any elements $v, w \in V$, we have $\Vert v + w \Vert \leq \Vert v \Vert + \Vert w \Vert$

##### Proof

It will suffice to prove that

$$
\Vert v + w \Vert^2 \leq (\Vert v \Vert + \Vert w \Vert)^2
$$

To do this, we observe that

$$
\begin{aligned}
\Vert v + w \Vert^2 &= \langle v + w, v + w \rangle \\
&= \langle v + w, v \rangle + \langle v + w, w \rangle \\
&= \langle v, v + w \rangle + \langle w, v + w \rangle \\
&= \langle v, v \rangle + \langle v, w \rangle + \langle w, v \rangle + \langle w, w \rangle
\end{aligned}
$$

But

$$
\begin{aligned}
\langle v, w \rangle + \langle w, v \rangle &= \langle v, w \rangle + \overline{\langle v, w \rangle} \\
\leq 2 \vert \langle v, v \rangle \vert
\end{aligned}
$$

Hence by [Schwarz](#Schwarz-inequality),

$$
\begin{aligned}
\Vert v + w \Vert^2 &\leq \Vert v \Vert^2 + 2 \vert \langle v, v \rangle \vert + \Vert w \Vert^2 \\
&\leq \Vert v \Vert^2 + 2 \Vert v \Vert \Vert w \Vert + \Vert w \Vert^2 \\
&= (\Vert v \Vert + \Vert w \Vert)^2
\end{aligned}
$$

Hence

$$
\sqrt{\Vert v + w \Vert^2} \leq \sqrt{(\Vert v \Vert + \Vert w \Vert)^2}
$$

Thus

$$
\Vert v + w \Vert \leq \Vert v \Vert + \Vert w \Vert
$$

Q.E.D.

### Orthonormal basis of a vector space over the complex field

An element $v$ of $V$ is said to be a unit vector as in the real case, if $\Vert v \Vert = 1$.

An orthogonal basis $\{v_1, ..., v_n \}$ is said to be **orthonormal** if it consists of unit vectors.

As before, we obtain an orthonormal basis from an orthogonal one by dividing each vector by its norm.

Let $\{e_1, ... ,e_n\}$ be an orthonormal basis of $V$. Let $v, w \in V$. There exist complex numbers $\alpha_1, ... ,\alpha_n \in C$ and $\beta_1, ..., \beta_n \in C$ such that

$$
v = \alpha_1 e_1 + ... + \alpha_n e_n
$$

and

$$
w = \beta_1 e_1 + ... + \beta_n e_n
$$

Then

$$
\begin{aligned}
\langle v, w \rangle &= \langle \alpha_1 e_1 + ... + \alpha_n e_n, \beta_1 e_1 + ... + \beta_n e_n \rangle \\
&= \sum_{i = 1}^n \sum_{j = 1}^n \alpha_i \overline{\beta_j} \langle e_i, e_j \rangle \\
&= \alpha_1 \overline{\beta_1} + ... + \alpha_n \overline{\beta_n}
\end{aligned}
$$


Thus in terms of this orthonormal basis, if $A, B$ are the coordinate vectors of $v$ and $w$ respectively, the hermitian product is given by the product described in [Example 3](#Example-3), namely $A \cdot \overline{B}$.

### Theorem 2.7

Let $V$ be either a vector space over $R$ with a positive definite scalar product, or a vector space over $C$ with a positive definite hermitian product. Assume that $V$ is finite dimensional. Let $W$ be a subspace of $V$. Then $V$ is the direct sum of $W$ and $W^\perp$.
