Let $V$ be a finite dimensional vector space over field $K$, with a scalar product.

Recall by [theorem 2.3](./5_2_orthogonal_bases_positive_definite_case.ipynb#Theorem-2.3), if the scalar product is positive definite, and if $W$ is a subspace, then $V$ is the direct sum of $W$ and $W^\perp$. **(1)**

However in the case of non positive definite scalar product, the argument in the proof in no longer value, as it is possible that for a vector $v$, $\langle v, v\rangle = 0$. For instance, over the complex numbers, $(1, i)$ is such a vector, as $(1, i) \cdot (1, i) = 1 \cdot 1 + i \cdot i = 1 - 1 = 0$.

And indeed the statement **(1)** is not generally true.

However the theorem concerning the existence of an orthogonal basis is still true, and we shall prove it by a suitable modification of the argument given by [the preceding section.](./5_2_orthogonal_bases_positive_definite_case.ipynb#Theorem-2.1) 

## Definitions: Null scalar product and null space

Given a scalar product of a vector space $V$, if $\langle u, u \rangle = 0$ for every $u \in V$, then the scalar product is said to be **null**, and $V$ is called a **null space**.

### Lemma 5.0: Every two elements in a null space are orthogonal under the null scalar product

Let $V$ be a vector space with a null scalar product. Then for any $v, w \in V$, $\langle v, w \rangle = 0$.

#### Proof

For any $v, w$ in $V$, we have

$$
\begin{aligned}
\frac{1}{2}(\langle v + w, v + w \rangle - \langle v, v \rangle - \langle w, w \rangle) &= \frac{1}{2} (\underbrace{(\langle v + w, v \rangle + \langle v + w, w \rangle)}_{\text{SP 2}} - \langle v, v \rangle - \langle w, w \rangle) \\
&= \frac{1}{2} ((\underbrace{\langle v, v + w \rangle}_{\text{SP 1}} + \underbrace{\langle w,  v + w \rangle}_{\text{SP 1}}) - \langle v, v \rangle - \langle w, w \rangle) \\
&= \frac{1}{2} (((\underbrace{\langle v, v \rangle + \langle v, w \rangle}_{\text{SP 2}}) + (\underbrace{\langle w,  v \rangle + \langle w,  w \rangle}_{\text{SP 2}})) - \langle v, v \rangle - \langle w, w \rangle) \\
&= \frac{1}{2}(\langle v, w \rangle + \langle w, v \rangle) \\
&= \frac{1}{2}(\langle v, w \rangle + \underbrace{\langle v, w \rangle}_{\text{SP 1}}) \\
&= \langle v, w \rangle
\end{aligned}
$$

And since the scalar product is null, we have

$$
\begin{array}{lll}
\langle v, v \rangle &= 0 \\
\langle w, w \rangle &= 0 \\
\langle v + w, v + w \rangle &= 0 \\
\end{array}
$$

Thus

$$
\frac{1}{2}(\langle v + w, v + w \rangle - \langle v, v \rangle - \langle w, w \rangle)
$$

Hence $\langle v, w \rangle = 0$, for any $v, w$ in $V$.

As a consequence, any basis of $V$ is an orthogonal basis.

## Theorem 5.1.

Let $V$ be a finite dimensional vector space over the field $K$, and assume that $V$ has a scalar product. If $V \neq \{\mathit{0}\}$, then $V$ has an orthogonal basis.

### Proof

#### Case 1: One-dimension

Assume $\dim V = 1$. Then any non-zero element $v \in V$ can form a basis as $\{v\}$, which is an orthogonal basis as it has only one element.

#### Case 2: $n$-dimensions ($n > 1$)

Assume $\dim V > 1$.

##### Case 2.1. Null space

Assume the scalar product is null. Thus, for every element in $v \in V$, we have $\langle v, v \rangle = 0$. Then by [lemma 5.0](#Lemma-5.0:-Every-two-elements-in-a-null-space-are-orthogonal-under-the-null-scalar-product), for any $v, w \in V$, $\langle v, w \rangle = 0$. Hence any basis in $V$ is an orthogonal basis.

##### Case 2.1. Non-null space

Assume the scalar product is *not* null. Hence there exists an element $v_1 \in V$ such that $\langle v_1, v_1 \rangle \neq 0$.

Inspired by the [Gram-Schmidt orthogonalization](./5_2_orthogonal_bases_positive_definite_case.ipynb#Proof), we let $v$ be any element in $V$, and for each $v$, we define $c$ as

$$
c = \frac{\langle v, v_1 \rangle }{\langle v_1, v_1 \rangle}
$$

Let $V_1$ be the $1$-dimensional vector space generated by $v_1$. Then [we know](./5_1_scalar_products.ipynb#Mutual-orthogonal-set-and-its-component-along-the-elements-of-the-set) $\langle v_1, v - c v_1 \rangle = 0$. Hence $\langle v_1, v - c v_1 \rangle \in V_1^{\perp}$.

Meanwhile, for every $v \in V$, we can write

$$
v = (v - c v_1) + c v_1
$$

Recall $V_1^\perp$ [is a subspace](./5_1_scalar_products.ipynb#Definition:-Orthogonal-space) of $V$. Hence $V$ is the [sum](../1_vector_spaces.ipynb#A-sum-is-a-subspace) of $V_1$ and $V_1^\perp$.

Let $v_2 \in V_1 \cap V^\perp$. Then we know there exists $c^\prime \in K$ such that $v_2 = c^\prime v_1$ (as $\dim V_1 = 1$), and also $\langle v_2, v_1 \rangle = 0$. Hence we have $\langle c^\prime v_1, v_1 \rangle = c^\prime \langle v_1, v_1 \rangle = 0$. Since $\langle v_1, v_1 \rangle \neq 0$, we have $c^\prime = 0$. Thus $V_1 \cap V^\perp = \{\mathit{0}\}$.

Then from [theorem 4.1 in chapter 1](../1_vector_spaces/1_vector_spaces.ipynb#Theorem-4.1.), we know $V = V_1 \oplus V_1^\perp$, thus $V$ is a direct sum of $V_1$ and $V^\perp$.

Then by [theorem 4.3 in chapter 1](../1_vector_spaces/1_vector_spaces.ipynb#Theorem-4.3.), we have

$$
\dim V_1^\perp = \dim V - \dim V_1 = \dim V - 1.
$$

Then we can by induction repeat the process to find $v_2 \in V_1^\perp$ that is orthogonal to $\{v_1\}$, and $v_3 \in (V_1^\perp)^\perp$, ... and finally $v_n$ that is orthogonal to $\{v_1, v_2, v_3 ..., v_{n-1}\}$. And since $\langle v_i, v_i \rangle \neq 0$ for $i \in 1, ..., n$, we know $v_i \neq \mathit{0}$. Then by [lemma 2.8](./5_2_orthogonal_bases_positive_definite_case.ipynb#Lemma-2.8) $\{v_1, ..., v_n\}$ is an orthogonal basis of $V$.

Q.E.D.

### Example 1. 

In $R^2$, let $X = \begin{pmatrix} x_1 \\ x_2 \end{pmatrix}$, and $Y = \begin{pmatrix} y_1 \\ y_2 \end{pmatrix}$. Define their product

$$
\langle X, Y \rangle = x_1 y_1 - x_2 y_2
$$

With this definition of scalar product, we have $\langle \begin{pmatrix} 1 \\ 0 \end{pmatrix}, \begin{pmatrix} 0 \\ 1 \end{pmatrix} \rangle = 1 \cdot 0 - 0 \cdot 1$ = 0. Thus $\{\begin{pmatrix} 1 \\ 0 \end{pmatrix}, \begin{pmatrix} 0 \\ 1 \end{pmatrix} \}$ is still an orthogonal basis.

For $\begin{pmatrix} 1 \\ 2 \end{pmatrix}$ and $\begin{pmatrix} 2 \\ 1 \end{pmatrix}$, we have $\langle \begin{pmatrix} 1 \\ 2 \end{pmatrix}, \begin{pmatrix} 2 \\ 1 \end{pmatrix} \rangle = 1 \cdot 2 - 2 \cdot 1 = 0$. Thus $\{ \begin{pmatrix} 1 \\ 2 \end{pmatrix}, \begin{pmatrix} 2 \\ 1 \end{pmatrix} \}$ is an orthogonal basis with this definition of scalar product. However it is not an orthogonal basis under the definition of the ordinary scalar product.
