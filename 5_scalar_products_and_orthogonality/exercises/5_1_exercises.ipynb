## 1. Let $V$ be a vector space with a scalar product. Show that $\langle \mathit{0}, v \rangle = 0$ for all $v$ in $V$.

### Proof

We [have already proven](../5_1_scalar_products.ipynb#Lemma-1.0) $\langle v, \mathit{0} \rangle = 0$ for all $v$ in $V$.

Then by **SP 1**, we have $\langle \mathit{0}, v \rangle = \langle v, \mathit{0} \rangle = 0$. Q.E.D.

## 2. Assume that the scalar product is positive definite. Let $v_1, ..., v_n$ be non-zero elements which are mutually orthogonal, that is $\langle v_i, v_j \rangle = 0$ if $i \neq j$. Show that they are linearly independent.

### Proof

Let $a_1, ..., a_n$ be some number such that

$$
a_1 v_1 + ... a_n v_n = \mathit{0}
$$

Then for each $j$ in $1, ..., n$ such that by **SP 2** and **SP 3**,

$$
\begin{aligned}
0 &= \langle v_j, \mathit{0} \rangle \\
&= \langle v_j, a_1 v_1 + ... a_n v_n \rangle \\
&= a_1 \langle v_j, v_1 \rangle + ... + a_j \langle v_j, v_j \rangle + ... + a_n \langle v_j, v_n \rangle
\end{aligned}
$$

Thus

$$
a_1 \langle v_j, v_1 \rangle + ... + a_j \langle v_j, v_j \rangle + ... + a_n \langle v_j, v_n \rangle = 0
$$

And since the scalar product is positive definite, then

$$
\langle v_j, v_j \rangle > 0
$$

Thus

$$
a_j = 0
$$

Hence for each $j$ in $1, ..., n$, we have $a_j = 0$.

Hence $v_1, ..., v_n$ are linear independent. Q.E.D.

## 3.

Let $M$ be a square $n \times n$ matrix which is equal to its transpose. If $X, Y$ are column $n$-vectors, then

$$
^t X M Y
$$

is a $1 \times 1$ matrix, which we identify with a number. Show that the map

$$
(X, Y) \mapsto ^t X M Y
$$

satisfies the three properties **SP 1, SP 2, SP 3** (hence is a scalar product). Give an example of a $2 \times 2$ matrix $M$ such that the product is not positive definite.

### Proof: the map is a scalar product

#### Proof: SP 1

We have

$$
(Y, X) \mapsto ^t Y M X
$$


By [theorem 3.3 in chapter 2.3](../../2_matrices/2_3_multiplication_of_matrices.ipynb#Theorem-3.3:-Multiplication-of-transposed-matrices), we have

$$
\begin{aligned}
^t Y M X &= ^t Y ^t (^t M) ^t (^t X) \\
&= ^t(^t X ^t M Y)
\end{aligned}
$$

And since $M = ^t M$, we have

$$
^t(^t X ^t M Y) = ^t(^t X M Y)
$$

And since $^t X M Y$ is an $1 \times 1$ matrix, we have $ ^t(^t X M Y) = ^t X M Y$.

Thus $^t X M Y = ^t Y M X$.

#### Proof: SP 2

Let $Z$ be another $n$-column vector. Then we have

$$
^t X M (Y + Z) = ^t X M Y + ^t X M Z
$$

#### Proof: SP 3

Let $c$ be a number. Then we have

$$
^t X M (c Y) = c (^t X M Y)
$$

and

$$
^t (c X) M Y = c (^t X M Y)
$$

Thus $(X, Y) \mapsto ^t X M Y$ is a scalar product. Q.E.D.

### Example:  an $M$ such that $(X, Y) \mapsto ^t X M Y$ is not positive definite

$$
M = \begin{pmatrix} 0 & 0 \\ 0 & 0 \end{pmatrix}
$$
