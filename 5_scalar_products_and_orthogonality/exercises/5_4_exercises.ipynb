## 1. 

Let $A$ be an $n \times n$ matrix, and assume that $A$ is symmetric, i.e. $A = ^t A$ Let $g_A: K^n \times K^n \to K$ be its associated bilinear map. Show that

$$
g_A(X, Y) = g_A(Y, X)
$$

for all $X, Y \in K^n$, and thus that $g_A$ is a scalar product, i.e. satisfies conditions [**SP 1**, **SP 2**, and **SP 3**](../5_1_scalar_products.ipynb#Definition:-Scalar-Products).

### Proof

#### Proof: $g_A(X, Y) = g_A(Y, X)$

By [theorem 3.3 in chapter 2.3](../../2_3_multiplication_of_matrices.ipynb#Theorem-3.3:-Multiplication-of-transposed-matrices), $^t (A B) = ^t B ^t A$. Then

$$
\begin{aligned}
g_A(X, Y) &= ^t X A Y \\
&= ^t X (A Y) \\
&= ^t(^t (^t X (A Y))) \\
&= ^t (^t (A Y) X) & \text{by theorem 3.3 in chapter 2.3}  \\
&= ^t ((^t Y ^t A) X) \\
&= ^t ((^t Y A) X) &\text{as $^t A = A$} \\
&= ^t (^t Y A X) \\
&= ^t Y A X &\text{as the product is a number in $F$} \\
&= g_A(Y, X)
\end{aligned}
$$

#### Proof: $g_A$ is a scalar product

- **SP 1**: We have already proven $g_A(X, Y) = g_A(Y, X)$ for any $X, Y \in K^n$.
- **SP 2**: Is proven since $g_A$ is a bilinear map.
- **SP 3**: Is proven since $g_A$ is a bilinear map.

Thus $g_A$ is a scalar product. Q.E.D.

## 2.

Conversely, assume that $A$ is an $n \times n$ matrix such that

$$
g_A(X, Y) = g_A(Y, X)
$$

for all $X, Y$. Show that A is symmetric.

### Proof

$$
g_A(X, Y) = ^t X A Y
$$

$$
g_A(Y, X) =  ^t Y A X
$$

Since

$$
g_A(X, Y) = g_A(Y, X)
$$


thus

$$
^t X A Y = ^t Y A X
$$

Hence

$$
\begin{aligned}
^t X A Y &= ^t (^t (^t X A Y)) \\
&= ^t (^t (^t X A Y)) \\
&= ^t (^t (^t X (A Y))) \\
&= ^t (^t (A Y) X) \\
&= ^t (^t Y ^t A X) \\
&= ^t Y ^t A X &\text{as the product is a number in $F$} \\
&= ^t Y A X
\end{aligned}
$$

Thus

$$
^t Y ^t A X  = ^t Y A X
$$

Hence $g_A = g_{^t A}$.

Then by [theorem 4.1](../5_4_bilinear_maps_and_matrices.ipynb#Theorem-4.1.), $A = ^t A$. Q.E.D.

## 3. Show that the bilinear maps of $K^n \times K^m$ into $K$ form a vector space. More generally, let $\operatorname{Bil}(U \times V, W)$ be the set of bilinear maps of $U \times V$ into $W$. Show that $\operatorname{Bil}(U \times V, W)$ is a vector space.

### Proof

#### Proof: $\operatorname{Bil}(K^n \times K^m, K)$ is a vector space

[Already proven](../5_4_bilinear_maps_and_matrices.ipynb#Proof:-$\operatorname{Bil}(K^m-\times-K^n,-K)$-is-a-vector-space).

#### Proof: $\operatorname{Bil}(U \times V, W)$ is a vector space

Since $W$ is a vector space, we [know](../../1_vector_spaces/1_vector_spaces.ipynb#Generalization:-vector-value-function-space-is-a-vector-space) that the space of all functions $g: U \times V \to W$ is a vector space.

Now we contend $\operatorname{Bil}(U \times V, W)$ is a subspace of the vector space of all functions $g: U \times V \to W$.

- **(i)**: Let $f: U \times V to W, g: U \times V to W$ be two bilinear maps in $\operatorname{Bil}(U \times V, W)$. 

Let $Y_1, Y_2$ be any vectors in $V$. Then for any fixed $X$,

$$
\begin{aligned}
(f + g)(X, Y_1 + Y_2) &= f(X, Y_1 + Y_2) + g(X, Y_1 + Y_2) \\
&= f(X, Y_1) + f(X, Y_2) + g(X, Y_1) + g(X, Y_2) &\text{as $f, g$ are bilinear maps} \\
&= (f + g)(X, Y_1) + (f + g)(X, Y_2)
\end{aligned}
$$

And let $c$ be any number in $K$. Then for any fixed $X$,

$$
\begin{aligned}
f(X, c Y) = c f(X, Y) &\text{as $f, g$ are bilinear maps}
\end{aligned}
$$

Hence for any fixed $X$, the mapping $Y \mapsto (f + g)(X, Y)$ is a linear map.

And similarly, we can prove for any fixed $Y$, the mapping $X \mapsto (f + g)(X, Y)$ is a linear map.

Hence $f + g$ is a bilinear map in $\operatorname{Bil}(U \times V, W)$.

- **(ii)**: Let $f$ be any bilinear map in $\operatorname{Bil}(U \times V, W)$. 

Let $c$ be any number in $K$.

Let $Y_1, Y_2$ be any vectors in $V$. Then for any fixed $X$,

$$
\begin{aligned}
(c f)(X, Y_1 + Y_2) &= c (f(X, Y_1 + Y_2)) \\
&= c (f(X, Y_1) + f(X, Y_2)) &\text{as $f$ is a bilinear map} \\
&= c f(X, Y_1) + c f(X, Y_2)
\end{aligned}
$$

And let $d$ be any number in $K$. Then for any fixed $X$,

$$
\begin{aligned}
(c f)(X, d Y) &= c (f(X, d Y)) \\
&= c (d f(X, Y)) &\text{as $f$ is a bilinear map} \\
&= d (c f)(X, Y)
\end{aligned}
$$

Hence for any fixed $X$, the mapping $Y \mapsto (c f)(X, Y)$ is a linear map.

And similarly, we can prove for any fixed $Y$, the mapping $X \mapsto (c f)(X, Y)$ is a linear map.

Hence $c f$ is a bilinear map in $\operatorname{Bil}(U \times V, W)$.

- **(iii)**: It's obvious that $\mathit{0}: U \times V, W$ defined as $\underbrace{\mathit{0}}_{\text{mapping}}(X, Y) = \underbrace{\mathit{0}}_{\text{vector in $W$}}$ is a bilinear map in $\operatorname{Bil}(U \times V, W)$.

Hence $\operatorname{Bil}(U \times V, W)$ is subspace of the space of all functions $f: U \times V, W$. Hence $\operatorname{Bil}(U \times V, W)$ is a vector space.
