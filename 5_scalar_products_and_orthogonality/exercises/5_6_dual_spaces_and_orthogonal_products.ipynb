## 1.

Let $A, B$ be two linearly independent vectors in $R^n$. What is the dimension of the space perpendicular to both $A$ and $B$?

### Solution

By [theorem 5.6](../5_6_dual_space_and_scalar_products.ipynb#Theorem-6.4:-The-dimension-of-a-vector-space-is-the-sum-of-the-dimension-of-a-subspace-and-the-dimension-of-its-orthogonal-complement) the dimension of the space perpendicular to both $A$ and $B$ is $n - 2$.

## 4.

Let $V$ be a vector space of finite dimension $n$ over the field $K$. Let $\varphi$ be a functional on $V$, and assume $\varphi \neq \mathit{0}$. What is the dimension of the kernel of $\varphi$? Proof?

### Solution and proof:

Since $\varphi$ is a functional on $V$, $\varphi: V \to K$ is a linear map. And since $\varphi \neq \mathit{0}$, we have $\operatorname{Im} \varphi = K$. Then by [theorem 3.2 in chapter 3.3](../../3_linear_mappings/3_3_kernel_and_image_of_a_linear_map.ipynb#Theorem-3.2:-Kernel,-image,-and-dimensions)

$$
\dim \operatorname{Ker} \varphi = \dim V - \dim \operatorname{Im} \varphi = n - 1.
$$

Q.E.D.

## 5.

Let $V$ be a vector space of dimension $n$ over the field $K$. Let $\psi, \varphi$ be two non-zero functionals on $V$. Assume that there is no element $c \in K, c \neq \mathit{0}$ such that $\psi = c \varphi$. Show that

$$
\dim (\operatorname{Ker} \varphi) \cap (\operatorname{Ker} \psi) = n - 2.

### 5.1: A relaxed version of the problem

Let $V$ be a vector space of dimension $n$ over the field $K$ with a non-degenerate scalar product. Let $\psi, \varphi$ be two non-zero functionals on $V$. Assume that there is no element $c \in K, c \neq \mathit{0}$ such that $\psi = c \varphi$. Show that


#### Proof

We know [the intersection of two subspace is also a subspace](../../1_vector_spaces/exercise/1.1_exercise.ipynb#10.-If-$U,-W$-are-subspaces-of-a-vector-space-$V$,-show-that-$U-\cap-W$-and-$U-+-W$-are-subspaces.). Thus $(\operatorname{Ker} \varphi) \cap (\operatorname{Ker} \psi)$ is a subspace of $V$.

And since $V$ has a non-degenerate scalar product, then the mapping

$$
v \mapsto L_v
$$

where

$$
L_v(w) = \langle v, w \rangle
$$

[give an isomorphism](../5_6_dual_space_and_scalar_products.ipynb#Theorem-6.2:-An-isomorphism-between-$V$-and-$V^*$) between $V$ and $V^*$. Thus we can find two unique vectors $u, w \in V$ such that $\varphi = L_u = \angle$ and $\psi = L_w$.

And since $\psi, \varphi$ be two non-zero functionals on $V$, and there is no element $c \in K, c \neq \mathit{0}$, we know $u, w$ are linearly independent.

Let $W$ be the vector space generated by $u$ and $w$.

Let $v$ be any vector in $(\operatorname{Ker} \varphi) \cap (\operatorname{Ker} \psi)$. Thus $L_u(v) = \varphi(v) = 0$ and $L_w(v) = \psi(v) = 0$. Hence $\langle v, u \rangle = 0$ and $\langle v, w \rangle = 0$. Hence $v \in W^\perp$. Hence $(\operatorname{Ker} \varphi) \cap (\operatorname{Ker} \psi) \subset W^\perp$.


Let $v$ be any vector in $W^\perp$. Then we have $\langle v, \langle u \rangle = 0$ and $\langle v, \langle w \rangle = 0$. Hence $\varphi(v) = L_u(v) = 0$ and $\psi(v) = L_w(u) = 0$. Hence $v \in (\operatorname{Ker} \varphi) \cap (\operatorname{Ker} \psi)$. Hence $W^\perp \subset (\operatorname{Ker} \varphi) \cap (\operatorname{Ker} \psi)$.

Hence $(\operatorname{Ker} \varphi) \cap (\operatorname{Ker} \psi) = W^\perp$. By [theorem 6.4](../5_6_dual_space_and_scalar_products.ipynb#Theorem-6.4:-The-dimension-of-a-vector-space-is-the-sum-of-the-dimension-of-a-subspace-and-the-dimension-of-its-orthogonal-complement), we know $\dim W^\perp = \dim V - \dim W = n - 2$. Hence $\dim (\operatorname{Ker} \varphi) \cap (\operatorname{Ker} \psi) = n - 2$. Q.E.D.
