## Definition: determinants of $2 \times 2$ matrices

Let

$$
A = \begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
$$

be a $2 \times 2$ matrix in field $K$. We define its determinant to be

$$
\begin{vmatrix}
a & b \\
c & d
\end{vmatrix} = a d - b c.
$$

## Determinant as a function

The determinant can be viewed as a function of the matrix $A$. It can also be viewed as a function of its two columns. Then we can write the determinant as

$$
\operatorname{Det}(A)
$$

or

$$
\operatorname{Det}(A^1, A^2).
$$

### Properties of determinant function

1. As a function of the column vectors, the determinant is linear.

This means: let $b^\prime, d^\prime$ be two numbers. Then

$$
\operatorname{Det} \begin{pmatrix}
a & b + b^\prime \\
c & d + d^\prime
\end{pmatrix} = 
\operatorname{Det} \begin{pmatrix}
a & b \\
c & d
\end{pmatrix} +
\operatorname{Det} \begin{pmatrix}
a & b^\prime \\
c & d^\prime
\end{pmatrix}.
$$

*Proof*: By the definition of determinant, we have

$$
\begin{aligned}
\operatorname{Det} \begin{pmatrix}
a & b + b^\prime \\
c & d + d^\prime
\end{pmatrix}
&= \begin{vmatrix}
a & b + b^\prime \\
c & d + d^\prime
\end{vmatrix} \\
&= a (d + d^\prime) + (b + b^\prime) c \\
&= a d + a d^\prime + b c + b^\prime c \\
&= (a d + b c) + (a d^\prime + b^\prime c) \\
&= \begin{vmatrix}
a & b \\
c & d
\end{vmatrix} +
\begin{vmatrix}
a & b^\prime \\
c & d^\prime
\end{vmatrix} \\
&= \operatorname{Det} \begin{pmatrix}
a & b \\
c & d
\end{pmatrix} +
\begin{pmatrix}
a & b^\prime \\
c & d^\prime
\end{pmatrix}
\end{aligned}
$$
